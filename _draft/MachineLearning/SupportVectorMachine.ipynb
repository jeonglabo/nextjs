### サポートベクターマシン (SVM)

#### 1. n次元空間でのベクトルを用いたハイパープレーンの表現

ハイパープレーンとは「その空間の次元より1次元少ない部分空間」を指します。つまり、n次元空間でのハイパープレーンはn-1次元の部分空間を意味します。例えば、3次元空間ではハイパープレーンは2次元の面、2次元空間では1次元の直線になります。複雑な問題に取り組む前に、まず3次元と2次元におけるハイパープレーンの方程式を考え、それをn次元に一般化してみましょう。

![図1. 決定境界 (Decision Surface)](出典: Support Vector Machines without tears)

平面の方程式は以下のように表されます。法線ベクトル $\vec{v} = (a, b, c)$ が与えられたとき、原点からの距離が $d$ である平面の方程式は $ax + by + cz + d = 0$ となります。これを2次元に適用すると、直線の方程式は直線に垂直なベクトル $\vec{w} = (a, b)$ に対して原点からの距離が $d$ である場合、 $ax + by + d = 0$ となります。

これを少し変形すると、 $y = -\frac{a}{b}x - \frac{d}{b}$ であることが分かり、この1次関数の傾きがベクトル $\vec{w}$ に垂直であることが確認できます。従って、法線ベクトルと任意の実数を利用すれば、n次元空間におけるハイパープレーンを簡単に定義できます。

n次元の実数空間におけるハイパープレーンは以下のように表されます：

$$
\vec{w} \cdot \vec{x} + b = 0
$$

ここで、$\vec{w}$は法線ベクトル、$\vec{x}$はハイパープレーン上の任意の点、$b$は距離を調整するための定数です。

#### 2. SVMの核心概念: 最大マージン分類器

SVMは与えられたデータポイントを2つのクラスの間で最も広いマージンを持つハイパープレーンを見つけるアルゴリズムです。このマージンとは、2つのクラスのデータ点からハイパープレーンまでの距離のことを指します。SVMはこのマージンを最大化するハイパープレーンを選択し、新しいデータポイントがどのクラスに属するかを決定します。

SVMの主要なポイントは次の通りです：

1. **サポートベクター**: マージンに最も近い位置にあるデータポイントです。SVMのハイパープレーンはこれらのサポートベクターによって決定されます。
2. **マージンの最大化**: サポートベクターとハイパープレーンの間の距離を最大化することで、分類の一般化能力を高めます。
3. **最適化問題**: このプロセスは最適化問題として定式化され、SVMの数学的基礎を形成します。

#### 3. SVMの数学的定式化

SVMの目標は、次の目的関数を最適化することです：

$$
\text{minimize } \frac{1}{2} \|\vec{w}\|^2
$$

制約条件は次の通りです：

$$
y_i (\vec{w} \cdot \vec{x}_i + b) \geq 1, \quad \forall i
$$

ここで、$y_i$は各データポイントのラベルで、$+1$または$-1$の値を取ります。この問題はラグランジュ乗数法を使用して解かれ、最適なハイパープレーンとマージンが決定されます。

#### 4. カーネルトリック (Kernel Trick)

SVMは、線形的に分離が困難なデータを処理するためにカーネル関数を使用します。カーネル関数はデータを高次元空間にマッピングし、非線形な分類を可能にします。これにより、SVMは複雑なデータパターンも学習できるようになります。

代表的なカーネル関数は以下の通りです：

- **線形カーネル**: $\langle \vec{x}, \vec{x}' \rangle$
- **多項式カーネル**: $(\langle \vec{x}, \vec{x}' \rangle + c)^d$
- **ガウシアンRBFカーネル**: $\exp(-\gamma \|\vec{x} - \vec{x}'\|^2)$

これらのカーネル関数を使用することで、SVMは高次元空間でデータを線形に分離し、複雑な境界線を生成して分類性能を向上させることができます。

#### 結論

サポートベクターマシンは、データの構造を解析し、最大マージン分類を通じて高い分類性能を保証する強力な機械学習手法です。さまざまなカーネル関数を用いることで、線形分類だけでなく非線形分類問題にも適用できるため、SVMは依然として多くの分野で広く使用されています。


### ラグランジュ乗数法の理解

ラグランジュ乗数法は、最適化問題において最大または最小値を求めるための数学的手法です。この方法は、目的関数 $ f(x, y) $ と制約条件 $ g(x, y) = 0 $ に対して、新たな変数 $ \lambda $ を用いて以下の補助方程式を作成し、その補助方程式を全ての変数に対して偏微分した値がゼロになるような解を見つけることによって実現されます。

$$
L(x, y, \lambda) = f(x, y) - \lambda g(x, y)
$$

上記のように目的関数と制約条件から補助方程式を作成して問題を解ける理由は、制約条件を満たしながら目的関数を最大化または最小化する点では、目的関数の勾配（簡単に言えば傾き）と制約条件の勾配が平行になるからです。以下の図を見ながら詳しく説明していきます。

図2と図3では、$ f(x, y) $ が $ d_1 $ から $ d_3 $ まで変化できる場合、制約条件 $ g(x, y) = c $ があるとします。このとき、制約条件を満たしつつ $ f(x, y) $ が取りうる最大値は $ d_1 $ です。

図3をよく見ると、$ f(x, y) $ と $ g(x, y) = c $ が接点を形成している部分が、制約を満たす $ f(x, y) $ の最大値であることがわかります。この接点を見つける条件は、2つの曲線の傾きが接点で平行であることです。曲線の傾きは微分で求めることができ、変数が多い場合は偏微分、さらに詳しくは勾配（グラディエント）を使って計算できます。接点の値を求める条件は以下の通りです。

$$
\nabla_{x, y} f = \lambda \nabla_{x, y} g
$$

ここで、

$$
\nabla_{x, y} f = \left(\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}\right)
$$

$$
\nabla_{x, y} g = \left(\frac{\partial g}{\partial x}, \frac{\partial g}{\partial y}\right)
$$

一般に複数の制約条件がある場合には次のように拡張することができます。

$$
L(x_1, \ldots, x_n, \lambda_1, \ldots, \lambda_M) = f(x_1, \ldots, x_n) - \sum_{k=1}^{M} \lambda_k g_k(x_1, \ldots, x_n)
$$

この場合、次の式の解を求めることで最適化条件を見つけることができます。

$$
\nabla_{x_1, \ldots, x_n, \lambda_1, \ldots, \lambda_M} L(x_1, \ldots, x_n, \lambda_1, \ldots, \lambda_M) = 0
$$



### 3. サポートベクターマシン (SVM)
#### a. 判別規則 (Decision Rule)

判別規則とは、新しい入力 $\vec{u}$ に対してそのクラスが + か - かを決定する方法のことです。次のように2次元空間において、判別境界 (ハイパープレーン) と法線ベクトル $\vec{w}$ があるとします。

![図4. Decision boundaryとdecision rule](図出典: Support Vector Machines without Tears)

法線ベクトル $\vec{w}$ の正確な値はまだわかりませんが、中心の破線よりも上にあれば + クラス、下にあれば - クラスとして分類することを目指します。これにより判別規則を次のように設定できます。

$$
\vec{w} \cdot \vec{u} \geq c
$$

ここで $ c $ は任意の定数です。この条件を満たす場合に + クラスとする判別規則を設定できます。この式を少し変形すると次のようになります。

$$
\vec{w} \cdot \vec{u} + b \geq 0
$$

この式は2次元での直線を意味しますが、これはハイパープレーンとして一般化できます。次に知るべきは、$\vec{w}$ と $ b $ の値です。

#### b. サポートベクターとストリートの生成条件

トレーニングの話に移りましょう。サンプルが2つのクラスに分かれているとします。1つは正のサンプル ($x^+$)、もう1つは負のサンプル ($x^-$) です。

式 (10) から正と負のサンプルに対して次のように拡張します。

次に図4を見ましょう。2つの実線上に正と負のサンプルが位置していることがわかります。

実線の間にある破線から実線までの距離は等しく設定されています。

このため、以下のように考えられます。式 (10) の破線は最も近い正のサンプルと負のサンプルから同じ距離 $\delta$ だけ離れている必要があり、他のサンプルは $\delta$ よりも離れていなければなりません。

次の2つの式を考えます。

$$
\vec{w} \cdot \vec{x}^+ + b \geq \delta
$$

$$
\vec{w} \cdot \vec{x}^- + b \leq -\delta
$$

ここで $\vec{w}$ と $ b $ の値はわからないので、任意の $\delta$ で正規化すると次のように表せます。

$$
\vec{w} \cdot \vec{x}^+ + b \geq 1
$$

$$
\vec{w} \cdot \vec{x}^- + b \leq -1
$$

これをさらに整理すると、サンプルのラベル $y_i$ を用いて次のようにまとめられます。

$$
y_i =
\begin{cases}
+1 & \text{for } x^+ \\
-1 & \text{for } x^-
\end{cases}
$$

これにより、次のように式 (13) と (14) を1つにまとめることができます。

$$
y_i (\vec{w} \cdot \vec{x}_i + b) \geq 1
$$

この式を少し変形すると、ハイパープレーンに最も近い位置にあるサンプルの条件は次のようになります。

$$
y_i (\vec{w} \cdot \vec{x}_i + b) - 1 = 0
$$

#### c. ストリートの幅と最適化問題の設定

図5では、実線上にある $x^+$ と $x^-$ のベクトル $\vec{x}^+$ と $\vec{x}^-$ について考えます。これらを用いて2つの実線間の距離を計算できます。

$$
\frac{\vec{w}}{|\vec{w}|} \cdot (\vec{x}^+ - \vec{x}^-)
$$

この式から、2つの実線間の距離は次のようになります。

$$
\frac{2}{|\vec{w}|}
$$

したがって、我々の目標はこの距離を最大化すること、すなわち $\frac{2}{|\vec{w}|}$ を最大化することです。数学的には、これを次のように最小化問題として表現することができます。

$$
\min \frac{1}{2} |\vec{w}|^2
$$

この設定により、サポートベクターマシンの最適化問題が定式化されます。


### d. ラグランジュ乗数法と制約条件の導出

したがって、私たちが直面する最適化問題は次のように定義されます。

**制約条件 (Constraints):**

$$
y_i (\vec{w} \cdot \vec{x}_i + b) - 1 = 0
$$

**目的関数 (Objective Function):**

$$
\frac{1}{2} |\vec{w}|^2
$$

これにより、次の補助方程式 $ L $ を立てることができます。

$$
L = \frac{1}{2} |\vec{w}|^2 - \sum_{i} \alpha_i [y_i (\vec{w} \cdot \vec{x}_i + b) - 1]
$$

ここで、補助方程式を各変数で偏微分します。

$$
\frac{\partial L}{\partial \vec{w}} = \vec{w} - \sum_{i} \alpha_i y_i \vec{x}_i = 0
$$

$$
\vec{w} = \sum_{i} \alpha_i y_i \vec{x}_i
$$

したがって、$\vec{w}$ はいくつかの $\vec{x}_i$ の線形結合 (linear sum) になります。「いくつかの」と言っているのは、$\alpha_i = 0$ の場合、そのサンプルは結合に含まれないためです。これは図4の実線上にないサンプルが、この線を決定するのに寄与しないことを意味します。

次に、この結果を最初の決定境界 (decision boundary) に代入すると、次の制約条件が得られます。

$$
\sum_{i} \alpha_i y_i \vec{x}_i \cdot \vec{u} + b = +1 \quad (\text{positive class})
$$

$$
\sum_{i} \alpha_i y_i \vec{x}_i \cdot \vec{u} + b = -1 \quad (\text{negative class})
$$

（ここで右辺が $\pm 1$ になるのは、実線上にないサンプルについての $\alpha$ の値が0になるためです。）

さらに、補助方程式 $ L $ を $ b $ で偏微分すると次のようになります。

$$
\frac{\partial L}{\partial b} = -\sum_{i} \alpha_i y_i = 0
$$

$$
\sum_{i} \alpha_i y_i = 0
$$

$\alpha_i$ に関しては、偏微分しても常に0になるため、$\alpha_i$ の値を求めるためには他の方法を用いる必要があります。これを求めるには、上記の制約条件を用いて連立方程式を解く必要があります。

次に、補助方程式 $ L $ に式 (31) と式 (35) を代入してみましょう。

$$
L = \frac{1}{2} |\vec{w}|^2 - \sum_{i} \alpha_i [y_i (\vec{w} \cdot \vec{x} + b) - 1]
$$

$$
= \frac{1}{2} \left(\sum_{i} \alpha_i y_i \vec{x}_i\right) \left(\sum_{j} \alpha_j y_j \vec{x}_j\right) - \sum_{i} \left(\alpha_i y_i \vec{w} \cdot \vec{x}_i + \alpha_i y_i b - \alpha_i\right)
$$

$$
= \sum_{i} \alpha_i - \frac{1}{2} \sum_{i} \sum_{j} \alpha_i \alpha_j y_i y_j \vec{x}_i \cdot \vec{x}_j
$$

### e. カーネルトリック (Kernel Tricks)

式 (41) では、補助方程式 $ L $ の最大値は $\vec{x}_i \cdot \vec{x}_j$ によって決まります。したがって、$\vec{x}_i \cdot \vec{x}_j$ を適切に変形することで、$ L $ の最大値をさらに引き上げることが可能です。

これを実現する方法の一つは、$\vec{x}_i$ に適切な変換 $\phi(\cdot)$ を施し、$\vec{x}_i \cdot \vec{x}_j$ の代わりに $\phi(\vec{x}_i) \cdot \phi(\vec{x}_j)$ を使用することです。

または、$\vec{x}_i \cdot \vec{x}_j$ に代わる適切な変換 $ K(\vec{x}_i, \vec{x}_j) $ を見つけ出し、代わりに使用することも可能です。

これを数式だけでなく、幾何学的に解釈すると以下のように説明できます。

![図6. カーネルの幾何学的な説明](出典: Support Vector Machines without Tears)

図6を見ると、左側の平面にあるサンプルが右側のように変形されていますが、これはサンプル空間をカーネル関数を使って3次元空間に変形させたためです。そして、3次元空間でハイパープレーンを使用することで、がんのサンプルと正常なサンプルをより簡単に区別できるようになります。

カーネルトリックによって、サンプルがより高次元の空間に投影され、複雑な境界を持つデータも線形分離可能になるため、SVMの強力な機能の一つとなっています。
