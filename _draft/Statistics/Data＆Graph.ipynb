##データの可視化

データの生成
まず、データを生成します。ここでは、ランダムなデータや特定の分布に従うデータを生成します。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 再現性のためのシード設定
np.random.seed(0)

# データポイント数
n_samples = 50

# 連続変数データの生成
x = np.linspace(0, 10, n_samples)
y_linear = 2 * x + 1 + np.random.normal(0, 5, n_samples)
y_quadratic = x**2 + np.random.normal(0, 10, n_samples)

# カテゴリカルデータの生成
categories = np.random.choice(['A', 'B', 'C'], size=n_samples)

# データフレームの作成
data = pd.DataFrame({
    'x': x,
    'y_linear': y_linear,
    'y_quadratic': y_quadratic,
    'category': categories
})
```

生成したデータを表形式でまとめます。以下の表は、生成された50サンプルのデータを示しています。

```python
import pandas as pd

# 生成されたデータフレームを表示
print(data.to_string(index=False))
```

###様々なグラフの描画

１．**散布図**：連続変数間の関係を視覚化します。

２．**回帰線付き散布図**：データのトレンドを把握できます。

３．**折れ線グラフ**：データの推移や変化を示します。

４．**ヒストグラム**：データの分布や偏りを確認できます。

５．**箱ひげ図**：データの中央値や四分位数、外れ値を示します。

６．**バイオリンプロット**：箱ひげ図と分布の形状を組み合わせて表示します。

７．**ヒートマップ**：相関関係を色で直感的に理解できます。

８．**ペアプロット**：複数の変数間の関係を一度に確認できます。

９．**棒グラフ**：カテゴリカルデータの頻度や割合を示します。

１０．**円グラフ**：全体に対する各カテゴリーの割合を視覚化します。

１１．**エリアチャート**：累積的なデータを示すのに適しています。

１２．**バブルチャート**：3つの変数を同時に表現できます（x, y, バブルサイズ）
。
１３．**ステムプロット**：データポイントのばらつきを示します。

１４．**誤差棒付き棒グラフ**：平均値とデータのばらつきを同時に示します。

１５．**カーネル密度推定プロット**：データの分布を滑らかに表示します。

以下では、50個のサンプルデータを生成し、それらを用いて様々な種類のグラフを描くPythonコードを提供します。

１．散布図

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['x'], data['y_linear'], color='blue', alpha=0.7)
plt.title('散布図: x vs y_linear')
plt.xlabel('x')
plt.ylabel('y_linear')
plt.grid(True)
plt.show()
```

２．回帰線付き散布図（Scatter Plot with Regression Line）

```python
sns.lmplot(x='x', y='y_linear', data=data, height=6, aspect=1.5)
plt.title('回帰線付き散布図')
plt.show()
```

３．折れ線グラフ（Line Plot）

```python
plt.figure(figsize=(8, 5))
plt.plot(data['x'], data['y_quadratic'], marker='o', linestyle='-', color='green')
plt.title('折れ線グラフ: x vs y_quadratic')
plt.xlabel('x')
plt.ylabel('y_quadratic')
plt.grid(True)
plt.show()
```

４．ヒストグラム（Histogram）

```python
plt.figure(figsize=(8, 5))
plt.hist(data['y_linear'], bins=15, color='purple', edgecolor='black', alpha=0.7)
plt.title('ヒストグラム: y_linearの分布')
plt.xlabel('y_linear')
plt.ylabel('頻度')
plt.grid(True)
plt.show()
```

５．箱ひげ図（Box Plot）

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='category', y='y_linear', data=data)
plt.title('箱ひげ図: カテゴリー別のy_linearの分布')
plt.xlabel('カテゴリー')
plt.ylabel('y_linear')
plt.show()
```

６．バイオリンプロット（Violin Plot）

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='category', y='y_quadratic', data=data, inner='quartile')
plt.title('Violin Plot: Distribution of   uadratic for each category')
plt.xlabel('Category')
plt.ylabel('y_quadratic')
plt.show()
```

７．ヒートマップ（Heatmap）

```python
# 相関行列の計算
corr = data[['x', 'y_linear', 'y_quadratic']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('相関ヒートマップ')
plt.show()
```

８．ペアプロット（Pair Plot）

```python
sns.pairplot(data, hue='category', height=2.5)
plt.show()
```

１１．エリアチャート（Area Chart）

１２．バブルチャート（Bubble Chart）

```python
plt.figure(figsize=(8, 5))
sizes = np.random.uniform(20, 200, size=n_samples)
plt.scatter(data['x'], data['y_quadratic'], s=sizes, alpha=0.5, c=sizes, cmap='viridis')
plt.title('バブルチャート: x vs y_quadratic')
plt.xlabel('x')
plt.ylabel('y_quadratic')
plt.colorbar(label='バブルサイズ')
plt.grid(True)
plt.show()
```

１３．ステムプロット（Stem Plot）

```python
plt.figure(figsize=(8, 5))
markerline, stemlines, baseline = plt.stem(data['x'], data['y_linear'], linefmt='grey', markerfmt='D', basefmt='black')
plt.title('ステムプロット: x vs y_linear')
plt.xlabel('x')
plt.ylabel('y_linear')
plt.grid(True)
plt.show()
```

１４．誤差棒付き棒グラフ（Bar Chart with Error Bars）

```python
means = data.groupby('category')['y_linear'].mean()
stds = data.groupby('category')['y_linear'].std()

plt.figure(figsize=(8, 6))
means.plot(kind='bar', yerr=stds, capsize=4, color='lightgreen', edgecolor='black')
plt.title('誤差棒付き棒グラフ: カテゴリー別のy_linearの平均')
plt.xlabel('カテゴリー')
plt.ylabel('y_linearの平均値')
plt.grid(True)
plt.show()
```

１５．カーネル密度推定プロット（Kernel Density Estimation Plot）

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data['y_quadratic'], shade=True, color='red')
plt.title('カーネル密度推定プロット: y_quadraticの分布')
plt.xlabel('y_quadratic')
plt.ylabel('密度')
plt.grid(True)
plt.show()
```
