### **反復測定分散分析（Repeated Measures ANOVA）**

反復測定分散分析（Repeated Measures ANOVA、RM ANOVA）は、同じ被験者が複数回の処置を受ける場合にデータを分析するための統計手法です。この手法は、複数の時点にわたって測定されたデータを比較する際に有効です。

#### **前提条件**
この内容を理解するために、以下の概念を知っていることが推奨されます：
- 標本と標準誤差の意味
- F値の意味と分散分析
- 対応標本t検定

### **ANOVAを理解するもう一つの観点**
分散分析で使用される「平方和（Sum of Squares, SS）」は非常に重要な概念です。平方和は偏差の平方を合計したもので、分散を推定する際に用いられる基本的な統計量です。平方和を使う理由は、偏差の絶対値を使用するよりも計算が簡便で、全体の平方和を特定の意味を持つ部分に分解できるため、分析において有用だからです。

### **用語の整理**
- **SS**：特定の条件によって説明される平方和
- **DF（自由度）**：統計的推定を行う際に独立して情報を提供するデータの数
- **MS（平均平方）**：SSを自由度で割った値

### **One-Way ANOVAをSSの観点から理解する**
分散分析は、すべての標本群が同一の母集団から抽出されたという帰無仮説のもとで行われます。ANOVAでは、2つの方法で分散を推定します：群内分散と群間分散です。群内分散に比べて群間平均の分散が大きい場合、帰無仮説は棄却されます。

F値は分散の比率として定義され、以下のように計算されます：

\[
F = \frac{s^2_{\text{between}}}{s^2_{\text{within}}}
\]

ここで、\(s^2_{\text{between}}\)は群間分散、\(s^2_{\text{within}}\)は群内分散を意味します。この比率により有意性を検証します。

### **RM ANOVAの計算過程**
RM ANOVAでは、全体の平方和が被験者間変動と被験者内変動に分けられます。被験者内変動はさらに処置による変動と残差変動に分けられます。反復測定データを分析する際に重視すべきなのは、時間経過による測定値の変化です。この変動を残差変動と比較してF値を計算します。

\[
F = \frac{\text{treatment SS}/\text{treatment DF}}{\text{residual SS}/\text{residual DF}} = \frac{\text{MS}_{\text{treat}}}{\text{MS}_{\text{res}}}
\]

このF値を用いて、時点間の統計的有意性を検証します。

### **球面性（Sphericity）**
球面性とは、すべての処置の組み合わせ間の差の分散が同じ場合を仮定するものです。球面性の仮定が満たされない場合、第1種の誤りが増加する可能性があるため、これを確認するためにMauchlyの検定を行います。球面性の仮定が満たされない場合は、自由度を補正して分析を続行します。
