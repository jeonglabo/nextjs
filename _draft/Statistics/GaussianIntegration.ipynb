## ガウス積分

ガウス積分は次のようにガウス関数に対する全実数範囲での積分で、その値は以下のようになります。

$$
\int_{-\infty}^{\infty} \exp(-x^2) \, dx = \sqrt{\pi} \quad \tag{1}
$$

### ガウス積分の計算過程
まず、以下のようにガウス積分の値を $ I $ とおきます。

$$
I = \int_{-\infty}^{\infty} \exp(-x^2) \, dx \quad \tag{2}
$$

このとき、 $ I $ を2乗した $ I^2 $ は次のように考えられます。

$$
I^2 = \left(\int_{-\infty}^{\infty} \exp(-x^2) \, dx\right) \left(\int_{-\infty}^{\infty} \exp(-y^2) \, dy\right) \quad \tag{3}
$$

$$
= \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \exp(-x^2) \exp(-y^2) \, dx \, dy \quad \tag{4}
$$

$$
= \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \exp(-(x^2 + y^2)) \, dx \, dy \quad \tag{5}
$$

ここで、直交座標系を極座標系に変換してみましょう。

$$
\begin{cases}
x = r\cos\theta \\
y = r\sin\theta
\end{cases} \quad \tag{6}
$$

### 直交座標系から極座標系への変換：積分範囲の修正
積分範囲は次のように変わります。

$$
\begin{cases}
-\infty < x < \infty \\
-\infty < y < \infty
\end{cases}
\Rightarrow
\begin{cases}
0 < r < \infty \\
0 < \theta < 2\pi
\end{cases} \quad \tag{7}
$$

式(7)のように積分範囲が変わる理由は、左側の $ x $ と $ y $ の範囲が実数全体を示しているため、極座標系で実数全体をカバーするには、半径 $ r $ は0から∞まで変化し、角度は0から $ 2\pi $ まで一周することでカバーできるからです。

### 直交座標系から極座標系への変換：微小面積の修正
式(5)にある微小面積 $ dx \, dy $ は以下のように変わります。

$$
dx \, dy \Rightarrow r \, dr \, d\theta \quad \tag{8}
$$

ここで、 $ dr $ は $ r $ 方向の微小変位、 $ d\theta $ は角度の微小変化を示しています。そのため、 $ r d\theta $ は弧の長さの微小変位と考えることができ、結果的に両者を掛け合わせると微小面積 $ r \, dr \, d\theta $ となります。

### 積分の続行
上記の2つの過程を考慮すると、式(5)の積分値は次のように表せます。

$$
\int_{0}^{2\pi} \int_{0}^{\infty} \exp(-r^2) \, r \, dr \, d\theta \quad \tag{9}
$$

この式は以下のようにも書けます。

$$
= \int_{0}^{2\pi} d\theta \int_{0}^{\infty} \exp(-r^2) \, r \, dr \quad \tag{10}
$$

$$
= 2\pi \int_{0}^{\infty} \exp(-r^2) \, r \, dr \quad \tag{11}
$$

ここで $ u = -r^2 $ と置換します。

$$
du = -2r \, dr \quad \tag{12}
$$

$$
r \, dr = -\frac{1}{2} du \quad \tag{13}
$$

したがって、式(11)は次のように書き換えられます。

$$
= 2\pi \int_{0}^{-\infty} -\frac{1}{2} \exp(u) \, du \quad \tag{14}
$$

$$
= 2\pi \cdot \left(-\frac{1}{2}\right) \cdot [\exp(u)]_{0}^{-\infty} \quad \tag{15}
$$

$$
= 2\pi \cdot \left(-\frac{1}{2}\right) \cdot (-1) = \pi \quad \tag{16}
$$

この結果は $ I^2 $ と同じであり、 $ I $ は常に正の値であるため、

$$
I = \int_{-\infty}^{\infty} \exp(-x^2) \, dx = \sqrt{\pi} \quad \tag{17}
$$

であることがわかります。
