### 検定統計量 (Test Statistic)
スチューデントのt検定を理解する前に、検定統計量という概念を一度押さえておくとよいでしょう。

「標本と標準誤差の意味」では、母集団、母数、標本、標本統計量について学び、最後に標本統計量は推定値であるため、推定誤差が発生し、それを標準誤差 (Standard Error) と呼ぶことを学びました。

では、検定統計量とは何でしょうか？

検定統計量は、私たちがよく「統計的に比較分析する」と言う際に使われる $ t $, $ F $, $ z $, $ \chi^2 $ などの概念を総称するもので、難しく考えずに「統計的仮説の真偽を検定するために」標本から計算する統計量です。

つまり、検定統計量は標本統計量を二次加工したもので、統計的仮説の真偽を検定するとは、検定統計量の値が基準を逸脱しているかを確認して、立てた仮説が間違っているかどうかを判断するプロセスです。

これだけではピンとこないかもしれませんが、今回はt値に焦点を当てて検定統計量について理解していきましょう。

### t値の意味：差異 / 不確実性
調査や研究を進める中で、2つの標本群間の差を比較する必要が生じることがあります。標本群間の差を比較する際には、いくつかの方法がありますが、主に比較される指標は平均です。

たとえば、新しい薬を開発した後、その薬の効果を確認するためにはどのような実験を計画できるでしょうか？

まず、臨床的に問題のない人々をできるだけ多く集め、彼らをランダムに2つのグループに分けます。一方のグループにはプラセボを、もう一方のグループには新しく開発した薬を投与し、その効果を確認します。

その結果、各グループで平均値を得ることができます。この2つの平均値の差を確認すれば、統計的に比較したことになるでしょうか？答えは「いいえ」です。これらの2つの平均値は標本平均であり、常に誤差を伴うことを忘れてはいけません。つまり、標本平均であるために発生する誤差を考慮しながら、2つの標本グループの平均差に関する指標を作成する必要があります。この誤差は標準誤差 (Standard Error) として知られています。

したがって、差を不確実性で割る方法で統計的な差の指標を作成します。

### 標本平均差の統計的指標：差異 / 不確実性
ここで言う不確実性は、2つの標本グループ間の平均差の不確実性を意味します。

t値の意味は、この「統計的指標」と同じです。つまり、t値は「これだけの差があって、この程度の誤差がある」ということを示します。

数式でt値を定義しましょう。2つの標本グループ間の差は $ \overline{X}_1 - \overline{X}_2 $ と書けます。前述した不確実性は2つの標本グループ間の平均差に関するものであり、数学的には標準誤差として表されますので、$ s_{\overline{X}_1 - \overline{X}_2} $ と書けます。

この「差の指標」にtという名前をつけると、次のように表せます。

$$
t = \frac{\overline{X}_1 - \overline{X}_2}{s_{\overline{X}_1 - \overline{X}_2}} \quad \tag{1}
$$

ここで、$ s_{\overline{X}_1 - \overline{X}_2} $ についてもう少し考えてみましょう。

$$
s_{\overline{X}_1 - \overline{X}_2} = \sqrt{ \text{Var}[\overline{X}_1 - \overline{X}_2] } \quad \tag{2}
$$

さらに展開すると、

$$
\text{Var}[\overline{X}_1 - \overline{X}_2] = \text{Var}[\overline{X}_1] + \text{Var}[\overline{X}_2] \quad \tag{4}
$$

$$
= \frac{\text{Var}[X_1]}{n_1} + \frac{\text{Var}[X_2]}{n_2} = \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} \quad \tag{6}
$$

ここで、式 (3) から (4) へ進む際には、2つのグループの平均値が独立して計算されていると考えることができるため、2グループ間の差に関する分散が各グループの分散の和として計算されました。また、式 (4) から (5) へ進む際には、次のように計算されています。

このようにして、t値を計算し、得られた結果が十分に大きければ、仮説検定の結果として「この2つの標本グループは同じ母集団から出たものではない」という結論を得ることができます。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.stats import ttest_ind, t, norm

from google.colab import drive
drive.mount('/content/drive')

# 設定
mksize = 15
XLIMs = [-5, 7]
YLIMs = [0.5, 6.5]
n_step = 100
sample_size = 6
np.random.seed(1)
data = np.random.normal(15, 2, 150)  # 仮のデータを生成
unique_data = np.unique(data)
my_order = []
for val in unique_data:
    indices = np.where(data == val)[0]
    my_order.extend(range(1, len(indices) + 1))

# t-分布の設定
xx_T = np.linspace(-5, 5, 100)
yy_T_orig = t.pdf(xx_T, df=sample_size * 2 - 2)
tvals2draw = []
my_max = 8.78

# アニメーションの作成
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

def update(frame):
    ax[0].cla()
    ax[1].cla()
    data_sort = np.sort(data)
    idx_perm1 = np.random.choice(len(data), sample_size, replace=False)
    idx_perm2 = np.random.choice(len(data), sample_size, replace=False)

    # 左のプロット
    for idx, val in enumerate(unique_data):
        indices = np.where(data == val)[0]
        ax[0].plot(data[indices], range(1, len(indices) + 1), 'o', markersize=mksize, markerfacecolor='0.8', markeredgecolor='k')
    ax[0].set_xlim([8, 22])
    ax[0].set_ylim([0, 40])
    ax[0].grid(True)
    ax[0].set_xlabel('height (cm)')
    ax[0].set_ylabel('count')
    ax[0].set_title('金星に住むエイリアン150人の身長')
    ax[0].plot(data_sort[idx_perm1], np.array(my_order)[idx_perm1], 'o', markersize=mksize, markerfacecolor='r', markeredgecolor='none')
    ax[0].plot(data_sort[idx_perm2], np.array(my_order)[idx_perm2], 'o', markersize=mksize, markerfacecolor='b', markeredgecolor='none')
    ax[0].legend(['サンプルグループ 1', 'サンプルグループ 2'])

    # 右のプロット
    t_stat, p_value = ttest_ind(data_sort[idx_perm1], data_sort[idx_perm2])
    t_stat = round(t_stat)
    y_coord = tvals2draw.count(t_stat) + 1
    ax[1].plot([t_stat] * y_coord, range(1, y_coord + 1), 'o', markersize=mksize, markerfacecolor='0.8', markeredgecolor='#D95319', linewidth=4)
    tvals2draw.append(t_stat)
    ax[1].set_xlim([-5, 5])
    ax[1].set_ylim([0, max(40, y_coord)])
    yy_T = yy_T_orig / np.max(yy_T_orig) * max(40, y_coord)
    ax[1].plot(xx_T, yy_T, color='b', linewidth=2)
    ax[1].grid(True)
    ax[1].set_xlabel('t-Value')
    ax[1].set_ylabel('Count')
    ax[1].set_title('n=6のサンプル集団2つを100回抽出し、そのたびのt値分布を確認')

# アニメーションの設定と保存
ani = FuncAnimation(fig, update, frames=n_step, repeat=False)
ani.save('/content/drive/MyDrive/Colab Notebooks/t_test_animation.gif', writer='imagemagick', fps=10)
plt.show()
```
