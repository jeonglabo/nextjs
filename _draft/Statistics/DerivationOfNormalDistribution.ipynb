##**正規分布の公式の導出**  


今回は、正規分布（またはガウス分布）の公式を導出してみます。

正規分布の公式は非常に複雑なので、以下の図のように3つのパートに分けて導出していきます。
image.png
**図1. 正規分布の公式と投稿での導出順序**

**前提条件**  
この投稿を理解するためには、以下の内容を知っていると良いです。

- 確率密度関数の概念と特性  
- ガウス積分  

###$e^{-x^2}$の形の導出

まずは、$f(x)$が$e^{-x^2}$の形を従うことを導出してみます。

####**必要な仮定**  
これを導くために、中心を直交座標系の原点に一致させた円形ダーツ盤にダーツを投げる過程を想像してみましょう。
image.png
**図2. 中心を原点に一致させた円形ダーツ盤**

**必要な仮定**  

1. ダーツ盤上で得点を等高線で表したとき、同じ得点の等高線に当たったダーツはすべて得点が同じです。つまり、確率密度関数は回転に対して独立です。
2. ダーツを投げて四角形の中に当てる状況を仮定したとき、目標とする四角形の面積が同じであれば、原点から四角形までの距離が近いほど、その四角形に当たる確率が高くなります。
3. 四角形までの距離が同じ場合、四角形の面積が広いほど当たる確率が高くなります。
image.png
**図3. （左）四角形の大きさが同じなら、距離が近いほど四角形に当たる確率が高い。つまり、四角形A、B、Cの順にダーツが当たる確率が高い。（右）四角形までの距離が同じ場合、四角形の面積が広いほど当たる確率が高い。つまり、F、E、Dの順に四角形に当たる確率が高い。**

###**導出過程**  
前述の3つの仮定を考慮し、直交座標上の任意の位置$(x, y)$で幅が$\Delta x$で高さが$\Delta y$の四角形Aにダーツが当たる期待値を考えてみましょう。
image.png






**図4. 期待値を計算する四角形A**

ここで、x軸とy軸でダーツが当たる確率の確率密度関数を$f(x, y)$とします。

このとき、x軸とy軸にダーツが当たる確率は互いに独立しているので、x軸とy軸に対する確率密度関数はそれぞれ$f(x)$と$f(y)$です。

したがって、四角形Aにダーツが当たる期待値は次のようになります。

$$
f(x) \Delta x f(y) \Delta y \quad \tag{1}
$$

また、上記の仮定のうち回転に関する仮定を利用するために、極座標系を利用して同じ確率密度を表現してみましょう。

極座標系で表現した確率密度関数を$g(r, \theta)$とすると、仮定1により、この確率密度関数は回転に対して独立しているため、$g(r, \theta) = g(r)$と表しても問題ありません。

したがって、四角形Aにダーツが当たる期待値を極座標系を使って表現すると以下のようになります。

$$
g(r) \Delta x \Delta y \quad \tag{2}
$$

ここで、式(1)と式(2)は同じ値なので、

$$
f(x) \Delta x f(y) \Delta y = g(r) \Delta x \Delta y \quad \tag{3}
$$

$\Delta x$と$\Delta y$を消去すると、

$$
f(x) f(y) = g(r) \quad \tag{4}
$$

になります。

ここで、仮定1をもう一度利用し、式(4)を$\theta$に関して微分しましょう。

**確率密度関数は回転に独立しているので、回転に関する微分の結果は0でなければなりません。**

$$
\frac{d f(x)}{d\theta} f(y) + f(x) \frac{d f(y)}{d\theta} = g(r) d\theta = 0 \quad \tag{5}
$$

この式は次のように書き換えることができます。

$$
\frac{d f(x)}{d x} \frac{d x}{d\theta} f(y) + f(x) \frac{d f(y)}{d y} \frac{d y}{d\theta} = g(r) d\theta = 0 \quad \tag{6}
$$

ここで、$x = r\cos(\theta)$、$y = r\sin(\theta)$ なので、

$$
\frac{d x}{d\theta} = -r\sin(\theta) \quad \tag{7}
$$

$$
\frac{d y}{d\theta} = r\cos(\theta) \quad \tag{8}
$$

となります。

したがって、式 (7) と式 (8) を式 (6) に代入すると、

$$
\frac{d f}{d x} (-r\sin(\theta)) f(y) + f(x) \frac{d f}{d y} (r\cos(\theta)) \quad \tag{9}
$$

ここで $ r\sin(\theta) = y $ であり、$ r\cos(\theta) = x $ なので、

$$
\frac{d f}{d x} (-y) f(y) + f(x) \frac{d f}{d y} x = 0 \quad  \tag{10}
$$

ここで、最初の項を右辺に移動して整理すると次のようになります。

視覚的な簡便さのため、$\frac{d f}{d x} = f'(x)$、$\frac{d f}{d y} = f'(y)$ と表記します。

$$
f(x) f'(y) x = f(y) f'(x) y \quad \tag{11}
$$

この式は常微分方程式であることが明らかです。変数分離法を使って解くために、両辺を $x$ と $y$ に関する式として表現してみましょう。

$$
x f(x) f'(x) = y f(y) f'(y) \quad \tag{12}
$$

式 (12) を見ると、両辺の分子と分母の比率が一定であることを意味していることがわかります。したがって、式 (12) の両辺の値はある定数 $C$ に等しいと言えます。

$$
x f(x) f'(x) = y f(y) f'(y) = C \quad \tag{13}
$$

これで、式 (13) から微分方程式を解くことにしましょう。$x$ や $y$ どちらで解いても同じ結果が得られるので、ここでは $x$ について解きます。

$$
x f(x) f'(x) = C \quad \tag{14}
$$

ここで、左辺に $x$ だけを残すように式を整理します。

$$
x = C \frac{f'(x)}{f(x)} \quad \tag{15}
$$

この両辺を積分すると、

$$
\frac{1}{2} x^2 = C \ln(f(x)) + C' \quad \tag{16}
$$

ここで $C'$ は積分によって生じた別の定数です。

したがって、$f(x)$ は次のように書けます。

$$
\therefore f(x) = A_0 \exp\left(\frac{1}{2} C x^2\right) \quad \tag{17}
$$

しかし、仮定2によると、標的の中央に近いほど当たる確率が高いため、式 (17) の指数部の値は負でなければなりません。

したがって、式 (17) を次のように書いて、指数部が負であることを強調します。

$$
f(x) = A_0 \exp\left(\frac{1}{2}(-k x^2)\right) \quad \text{where } k > 0 \quad \tag{18}
$$

**$\frac{1}{(\sigma \sqrt{2\pi})}$の導出**

次に、先に導出した式 (18) で $A_0$ の値が $\frac{1}{\sigma \sqrt{2\pi}}$ であることを導出します。

確率密度関数の特性を考えると、確率密度関数の全体の面積は1でなければなりません。

$$
\int_{-\infty}^{\infty} f(x) dx = 1 \quad \tag{19}
$$

したがって、次の式が成り立つ必要があります。

$$
\int_{-\infty}^{\infty} A_0 \exp\left(\frac{1}{2} - k x^2\right) dx = 1 \quad \tag{20}
$$

ここで、$A_0$は定数なので、

$$
\int_{-\infty}^{\infty} \exp\left(\frac{1}{2} - kx^2\right) dx = \frac{1}{A_0} \quad \tag{21}
$$

式 (21) の値を $I$ とすると、

$$
I^2 = \iint_{-\infty}^{\infty} \exp\left(-\frac{1}{2}k(x^2 + y^2)\right) dx dy \quad \tag{22}
$$

です。この二重積分の積分範囲を直交座標系から極座標系に変換すると、

$$
I^2 = \int_{\theta=0}^{2\pi} \int_{r=0}^{\infty} \exp\left(-\frac{1}{2}kr^2\right) r dr d\theta \quad \tag{23}
$$

となります。

ここで、次のように置換しましょう。

$$
-\frac{1}{2}kr^2 = u \quad \tag{24}
$$

すると、

$$
-k r dr = du \quad \tag{25}
$$

であり、

$$
r dr = -\frac{1}{k} du \quad \tag{26}
$$

となります。

したがって、式 (23) は次のように書けます。

$$
I^2 = \int_{\theta=0}^{2\pi} \int_{u=-\infty}^{0} \exp(u) \left(-\frac{1}{k} du\right) d\theta \quad \tag{27}
$$

$$
= -\frac{1}{k} \int_{\theta=0}^{2\pi} \int_{u=-\infty}^{0} \exp(u) du d\theta \quad \tag{28}
$$

$$
= -\frac{1}{k} \int_{\theta=0}^{2\pi} \left[\exp(u)\right]_{-\infty}^{0} d\theta \quad \tag{29}
$$

ここで、$\exp(-\infty) = 0$ であり、$\exp(0) = 1$ なので、

$$
= -\frac{1}{k} \int_{\theta=0}^{2\pi} (-1) d\theta \quad \tag{30}
$$

$$
= \frac{2\pi}{k} \quad \tag{31}
$$

となります。

したがって、この値は元の $I^2$ に等しいので、$I$ の値は次のようになります。

$$
I = \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}kx^2\right) dx = \sqrt{\frac{2\pi}{k}} \quad \tag{32}
$$

ここで、$I$ の値は確率密度関数の面積に関連するため常に正であるので、$I$ の値は正の値のみを持ちます。

また、式 (21) での $I$ の値は $\frac{1}{A_0}$ とも等しかったので、

$$
A_0 = \sqrt{\frac{k}{2\pi}} \quad \tag{33}
$$

となります。

したがって、式 (18) をもう一度書くと次のようになります。

$$
A_0 \exp\left(\frac{1}{2}(-kx^2)\right) = \sqrt{\frac{k}{2\pi}} \exp\left(\frac{1}{2}(-kx^2)\right) \quad \tag{34}
$$

ここで、$k$ に関する答えを得て、$A_0 = \frac{1}{\sigma\sqrt{2\pi}}$ であることを示すために、以下の指数部分の導出を続けていきます。

###**指数部の内部の式の導出**  
図1の正規分布の公式にある指数部の内部の式を導出するには、確率密度関数のモーメントの概念を利用する必要があります。

難しいことではなく、次のように確率密度関数 $ f(x) $ が与えられたとき、平均と分散は次のように計算できます。

$$
\mu = \int_{-\infty}^{\infty} x f(x) dx \quad \tag{35}
$$

$$
\sigma^2 = \int_{-\infty}^{\infty} x^2 f(x) dx \quad \tag{36}
$$

ここでは、式 (34) まで求めた $ f(x) $ の式を利用して平均と分散を求めてみます。

$$
\mu = \int_{-\infty}^{\infty} x \sqrt{\frac{k}{2\pi}} \exp\left(-\frac{1}{2}kx^2\right) dx \quad \tag{37}
$$

式 (37) では、$ x $ 項は奇関数であり、$\exp\left(-\frac{1}{2}kx^2\right)$ 項は偶関数です。したがって、奇関数と偶関数の積は奇関数なので、式 (37) の結果は 0 となります。

次に、分散の値を計算してみます。

$$
\sigma^2 = \int_{-\infty}^{\infty} x^2 \sqrt{\frac{k}{2\pi}} \exp\left(-\frac{1}{2}kx^2\right) dx \quad \tag{38}
$$

$$
= \sqrt{\frac{k}{2\pi}} \int_{-\infty}^{\infty} x^2 \exp\left(-\frac{1}{2}kx^2\right) dx \quad \tag{39}
$$

ここで、式 (39) を次のように考えます。

$$
\sqrt{\frac{k}{2\pi}} \int_{-\infty}^{\infty} x \cdot x \exp\left(-\frac{1}{2}kx^2\right) dx \quad \tag{40}
$$

ここで、部分積分を利用して式 (40) を積分してみましょう。

$ x $ を $ u $ とし、$ x\exp\left(-\frac{1}{2}kx^2\right) $ を $ dv $ とします。

$$
\begin{cases}
u = x \\
du = 1
\end{cases} \quad \tag{41}
$$

$$
\begin{cases}
dv = x \exp\left(-\frac{1}{2}kx^2\right) \\
v = -\frac{1}{k} \exp\left(-\frac{1}{2}kx^2\right)
\end{cases} \quad \tag{42}
$$

となります。

したがって、式 (40) の積分値は次のようになります。

$$
\sqrt{\frac{k}{2\pi}} \left\{\left[x \cdot \left(-\frac{1}{k}\right) \exp\left(-\frac{1}{2}kx^2\right)\right]_{-\infty}^{\infty} + \frac{1}{k} \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}kx^2\right) dx\right\} \quad \tag{43}
$$

式 (43) の大括弧の中の項について、無限大において $ x $ は無限大に発散し、指数項は 0 に収束しますが、指数項が 0 に収束する速度のほうが速いです。これは負の無限大の場合も同様で、指数項が 0 に収束する速度のほうが速いです。したがって、大括弧内の項は結局 0 になります。

したがって、式 (43) は

$$
\sqrt{\frac{k}{2\pi}} \left\{\frac{1}{k} \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}kx^2\right) dx\right\} \quad \tag{44}
$$

となり、ここで式 (44) の大括弧内の値は式 (32) から分かる値です。

$$
\sqrt{\frac{k}{2\pi}} \left(\frac{1}{k}\right) \sqrt{\frac{2\pi}{k}} = \frac{1}{k} \quad \tag{45}
$$

そして、この値はもともとの分散 $\sigma^2$ だったので、

$$
\therefore k = \frac{1}{\sigma^2} \quad \tag{46}
$$

です。

再び式 (34) に $ k $ の値を代入すると、

$$
f(x) = \sqrt{\frac{k}{2\pi}} \exp\left(-\frac{1}{2}kx^2\right) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{x^2}{2\sigma^2}\right) \quad \tag{47}
$$

となり、この式は正規分布の式で、平均が 0 の場合を意味します。平均が $\mu$ の場合は、$ x $ を $ x - \mu $ に平行移動すればよいので、最終的な正規分布の公式は次のようになります。

$$
\frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right) \quad \tag{48}
$$
