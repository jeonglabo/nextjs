### **カイ二乗分布と検定**

#### **カイ二乗分布**
統計学を学び始める際、多くの新しい用語に出会うことが最も難しい部分かもしれません。カイ二乗分布もその一つで、ギリシャ文字χ（カイ）は英語で "chi" と書かれ、「カイ」と読みます。この分布は、自由度$ k $のカイ二乗分布は、$ k $個の独立した標準正規分布に従う確率変数の二乗和の分布として定義されます。

$$
Q = \sum_{i=1}^{k} X_i^2
$$

#### **カイ二乗分布のシミュレーション**
カイ二乗分布を理解するために、標準正規分布からランダムに変数を1つ選び、その値を二乗してヒストグラムにカウントしていくことを考えます。これを繰り返すと、カイ二乗分布の形状に近づくヒストグラムが得られます。

#### **様々な自由度におけるカイ二乗分布の形状**
カイ二乗分布は、変数を「二乗」しているため、正の値のみに存在します。また、足し合わせる変数が多くなると、中心極限定理により分布は正規分布に近づきます。

#### **カイ二乗分布の用途**
カイ二乗分布は、誤差（エラー）や偏差を分析する際に役立ちます。例えば、回帰分析などでモデルの誤差が正規分布に従うと仮定されるため、カイ二乗分布を用いて誤差が偶然生じたのかどうかを判断できます。

#### **カイ二乗検定**
カイ二乗検定は、適合度検定（goodness-of-fit test）やクロス集計分析（cross tabulation analysis）などで使用され、次の式でカイ二乗統計量を計算します：

$$
\chi^2 = \sum_{i} \frac{(O_i - E_i)^2}{E_i}
$$

ここで、$ O_i $は観測値、$ E_i $は期待値を示します。この統計量はピアソンのカイ二乗統計量として知られています。

### **適合度検定**
適合度検定は、一つの独立変数について、理論的に期待される頻度分布と観測された頻度分布を比較するために使用されます。独立変数はカテゴリカル変数（質的変数）であることがポイントです。

#### **例：**
ある5種類のキャンディの袋から100個のキャンディを取り出して、各種類のキャンディの数を数えたとします。各種類の期待値が20個であると仮定して、以下のようにカイ二乗統計量を計算します：

$$
\chi^2 = \frac{(17 - 20)^2}{20} + \frac{(16 - 20)^2}{20} + \cdots + \frac{(14 - 20)^2}{20} = 7.9
$$

自由度は4であり、この場合のカイ二乗値は臨界値よりも小さいため、観測された分布は期待された分布と有意な違いがないと結論付けられます。

### **クロス集計分析**
クロス集計分析は、複数のカテゴリカル変数の関係を検討するために使用されます。各セルの期待値は次の式で計算されます：

$$
\text{期待値} = \frac{\text{該当行の合計} \times \text{該当列の合計}}{\text{全データの総数}}
$$

行数を$ r $、列数を$ c $とした場合、自由度は$(r - 1) \times (c - 1)$です。

#### **カイ二乗検定の証明（スキップ可能）**
ピアソンのカイ二乗検定がカイ二乗分布に従うことを証明するために、中心極限定理と各セルの期待値を用いて証明するプロセスがありますが、これを理解するためにはさらに高度な統計学の知識が必要となります。詳細な証明過程については、専門的な文献やリソースを参考にしてください。
