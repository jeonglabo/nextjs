##1種エラーと2種エラー

統計学を学ぶ中で、1種および2種のエラーという新しい用語に出会います。これらは特に数値から意味を得るわけではなく、区別のために1種(type I)、2種(type II)という名前がつけられたものです。

1種エラーは、**帰無仮説が真であるのにそれを誤って棄却する際に発生するエラー**です。  
2種エラーは、**帰無仮説が偽であるのにそれを棄却しなかった際に発生するエラー**です。

これらの定義を理解するために、まず「帰無仮説」とは何かを正確に押さえておく必要があります。また、「誤って棄却する」とは何を意味するのかについても確認していきましょう。

### 事件と帰無仮説

1種および2種のエラーについて話す前に、これらの内容は「事件に関する仮説」に関連していることを理解するのが重要です。つまり、私たちがt検定や分散分析などを通じて行うのは仮説検証であるということです。

ただし、どのような仮説検定も100％確実ではありません。検定は確率に基づいているため、いつも誤った結論を出す可能性があります。

では、私たちはどのような事柄に対して仮説を検証しているのでしょうか？たとえば、「私が住んでいるアパートで火事が発生した」というような出来事（事件）です。

通常、社会では事件が起こった可能性に焦点を当てることが多いですが、統計学では逆に、事件が「起こらなかった」という仮説に焦点を当てることが多いです。なぜなら、統計学の主要な仮説検証方法は、「事件が起こらなかったと仮定し、その可能性が非常に低いことを証明することで間接的に事件が起こったことを示す」ためです。

このように、統計学は事件が「起こらなかった」という仮定に重きを置くため、その仮定には「帰無仮説」という名前がつけられています。帰無仮説は、何も起こらなかったことを仮定する仮説です。

### 1種エラーと2種エラーの例

**図1: 1種エラーと2種エラーの理解に最適な例：火災警報システム**

#### 1種エラー: False Alarm
1種エラーの定義は「帰無仮説が真であるのに誤って棄却するエラー」です。つまり、何も起こっていない（火事は起こっていない）にもかかわらず、火災警報が誤って鳴ることです。これを一言で表すと「False Alarm（誤警報）」と言えます。

#### 2種エラー: Miss
2種エラーの定義は「帰無仮説が偽であるのに棄却しないエラー」です。つまり、実際には火事が起こっているにもかかわらず、火災警報が鳴らずにそのまま過ぎてしまう場合です。これを一言で表すと「Miss（見逃し）」と言えます。

### エラーのまとめ

| エラーの種類 | 説明                     | 例                      |
|-------------|--------------------------|-------------------------|
| 1種エラー   | 帰無仮説が真で誤って棄却  | 火事がないのに警報が鳴る|
| 2種エラー   | 帰無仮説が偽で棄却しない | 火事があるのに警報が鳴らない|

### p値との関係

p値は1種エラーと関連があります。p値は「帰無仮説が正しいとした場合、その仮説が現実である確率」を意味します。つまり、p値は1種エラーを犯す確率と同じ意味を持ちます。

これらの理解を深めることで、統計学におけるエラーの意味や影響をより的確に把握することができます。
