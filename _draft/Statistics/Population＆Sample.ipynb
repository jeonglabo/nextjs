### **募集団と標本集団の概念**

統計学では、**募集団**と**標本集団**の概念が非常に重要です。募集団は、研究者が関心を持つ対象の全体集合を意味し、これは非常に抽象的であったり、直接確認することが難しい場合が多いです。例えば、「成人男性200人」という架空の募集団を想定し、彼らの統計的特性を把握してみましょう。



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# データ生成
np.random.seed(1)
mean_height = 175  # 平均身長
std_dev_height = 7  # 標準偏差
n_samples = 200  # サンプル数

# 身長データの生成
heights = np.random.normal(mean_height, std_dev_height, n_samples)

# データをデータフレームに格納
df = pd.DataFrame(heights, columns=['Height (cm)'])

# データの一部を表示（先頭10行）
print(df.head(10))

# ヒストグラムの表示
plt.figure(figsize=(8, 6))
plt.hist(heights, bins=15, color='skyblue', edgecolor='black')
plt.title('Height Distribution of Adult Males')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

募集団の特性を数値で表す**母数**は以下のようなものがあります：

- **平均**: 募集団の中心値
- **分散、標準偏差**: 募集団のデータがどれほど広がっているかを示す
- **分位数**: データの特定の位置の値（中央値、第1四分位数など）
- **母比率**: 特定の事象が発生する確率

母数を知ることで、募集団の全体の特性を理解するのに役立ちます。しかし、現実では募集団全体を調査することは不可能なため、**標本集団**を抽出して母数を推定する方法が使われます。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# データ生成
np.random.seed(2)
mean_height = 175  # 平均身長
std_dev_height = 7  # 標準偏差
n_samples = 200  # サンプル数

# 身長データの生成
heights = np.random.normal(mean_height, std_dev_height, n_samples)

# データをデータフレームに格納
df = pd.DataFrame(heights, columns=['Height (cm)'])

# データの一部を表示（先頭10行）
df_sample = df.head(10)

# ヒストグラムの表示
plt.figure(figsize=(8, 6))
plt.hist(heights, bins=15, color='skyblue', edgecolor='black')

# 中央値と分位点の計算
median = np.median(heights)
q25 = np.percentile(heights, 25)
q75 = np.percentile(heights, 75)

# 中央値と分位点をヒストグラムに表示
plt.axvline(median, color='red', linestyle='--', linewidth=1.5, label=f'Median: {median:.2f} cm')
plt.axvline(q25, color='green', linestyle='--', linewidth=1.5, label=f'25th Percentile: {q25:.2f} cm')
plt.axvline(q75, color='purple', linestyle='--', linewidth=1.5, label=f'75th Percentile: {q75:.2f} cm')

# グラフのラベル
plt.title('Height Distribution of Adult Males')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# データの一部を表示
df_sample
```

### **標本集団と標本統計量**

**標本**とは、募集団の部分集合であり、募集団から無作為に抽出されたデータのことです。例えば、成人男性200人の中から任意に10人を抽出した結果は標本集団になります。標本を何度も抽出すると、各標本は異なり、それを通じてさまざまな標本統計量を得ることができます。

標本統計量は母数を推定するための値で、主な標本統計量は以下の通りです：

- **標本平均**: 標本の平均値
- **標本標準偏差**: 標本のデータがどれほど広がっているかを示す
- **標本比率**: 標本内で特定の事象が発生した比率

しかし、標本統計量は抽出するたびに値が変わる可能性があるため、**誤差**が発生します。

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate population data (heights in cm)
np.random.seed(42)
population = np.random.normal(loc=175, scale=7, size=200)

# Sampling 10 samples repeatedly and calculate the mean of each sample
num_samples = 100
sample_size = 10
sample_means = [np.mean(np.random.choice(population, sample_size, replace=False)) for _ in range(num_samples)]

# Plotting the diagram of sample means
plt.figure(figsize=(8, 5))

# Plotting the sample means as a line plot
plt.subplot(1, 2, 1)
plt.plot(sample_means, marker='o', linestyle='-', color='blue')
plt.axhline(np.mean(population), color='red', linestyle='--', label='Population Mean')
plt.title('Sample Means (10 samples each)')
plt.xlabel('Sample Number')
plt.ylabel('Mean Height (cm)')
plt.grid(True)
plt.legend()

# Plotting the histogram of the sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(population), color='red', linestyle='--', label='Population Mean')
plt.axvline(np.median(sample_means), color='green', linestyle='-', label='Median of Sample Means')
plt.title('Histogram of Sample Means')
plt.xlabel('Mean Height (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```

### **標準誤差: 推定の不確実性**

**標準誤差**（Standard Error, SE）は、標本統計量の標準偏差を意味し、推定に伴う不確実性を数値化したものです。例えば、何度も標本を抽出して得られた標本平均の標準偏差が標準誤差となります。

標本平均の標準誤差は以下の式で計算できます：

$$
SEM = \frac{s}{\sqrt{n}}
$$

ここで $ s $ は標本標準偏差、$ n $ は標本のサイズです。標準誤差は標本サイズが大きくなるほど小さくなり、これは推定がより正確になることを意味します。

---



### **標準偏差と標準誤差の違い**

標準偏差と標準誤差は混同しやすい概念ですが、意味が異なります：

- **標準偏差**: 募集団のデータがどれほど広がっているかを示す値です。
- **標準誤差 (SEM)**: 標本平均の推定値に対する不確実性を示す値です。

そのため、データの特性を記述する際には標準偏差を使用することが望ましいです。標準誤差は推定値の正確さを示す指標として使用されるべきで、誤って使用すると、データが実際よりも変動が少ないように見えることがあります。

統計的検定ではSEMを使って推定の正確さを説明しますが、実際の募集団の特性を説明する際には標準偏差を使用するのが適切です。

### 標本平均の標準誤差の数式的証明

図4で見た標本平均の標準偏差、すなわち平均の標準誤差（Standard Error of Mean, SEM）を求めてみましょう。

1回に $ n $ 個の標本を抽出する実験を行ったとき、標本平均は次のように計算されます。

$$
\overline{X} = \frac{1}{n} \sum_{i=1}^{n} X_i \quad (1)
$$

ここで、 $ X_i $ は1回抽出した標本群の標本値を意味します。

求めたいのはこの標本平均 $ \overline{X} $ の標準偏差ですので、まず標本平均の分散値を計算してみると次のようになります。

$$
\text{Var}[\overline{X}] = \text{Var}\left[\frac{1}{n} \sum_{i=1}^{n} X_i\right] = \text{Var}\left[\frac{1}{n} (X_1 + X_2 + \cdots + X_n)\right] \quad (2)
$$

以下の (3)、(4) のような分散演算子の2つの性質により、

$$
\text{Var}[aX + b] = a^2 \text{Var}[X] \quad (3)
$$

$$
\text{Var}[X_1 + X_2] = \text{Var}[X_1] + \text{Var}[X_2] \quad (4)
$$

式 (2) は次のように書き換えられます。

$$
\text{式} (2) = \frac{1}{n^2} \text{Var}[X_1 + X_2 + \cdots + X_n] \quad (5)
$$

$$
= \frac{1}{n^2} \times n \times \text{Var}[X] = \frac{1}{n} \text{Var}[X] \quad (6)
$$

つまり、母分散を $ \sigma^2 $ とすれば、標本平均の標準偏差は

$$
\sqrt{\text{Var}[\overline{X}]} = \frac{1}{\sqrt{n}} \text{std}[X] = \frac{1}{\sqrt{n}} \sigma \quad (7)
$$

母分散の推定値を $ s^2 $ としたとき、標本平均の標準偏差は推定値を用いて

次のように書くことができます。

$$
SEM = \sqrt{\text{Var}[\overline{X}]} = \frac{s}{\sqrt{n}} \quad (8)
$$

### 注意：標準誤差と標準偏差は異なる概念である

標本平均の標準誤差（SEM）と標準偏差を混同して使用したり、誤って使用する場合があります。特に論文を読んでいると、このようなミスをしばしば見かけることがありますが、式 (8) で見たように SEM は標準偏差より常に小さいため、データを記述するときに結果がより良く見えるからです。

再度整理すると、標準偏差は母集団の分布がどれだけ広がっているかを説明する概念であり、SEM は平均の推定値に対する不確実性を数値化したものです。

通常、結果を見る立場としては母集団に関心があることが多いため、データに関して記述する際は標準偏差（またはそれに準ずる推定値）を使用して記述すべきです。
