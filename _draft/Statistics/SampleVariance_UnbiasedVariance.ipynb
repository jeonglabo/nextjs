### 標本分散は $ n $ ではなく $ n-1 $ で割る理由

統計学では、ある母数を推定するために推定量が使用されます。代表的な母数には、母平均、母分散、母比率などがあります。

理想的には、標本のサイズが大きくなるほど、標本から計算して推定する値が推定したい母数に近づくことが望まれます。そのため、理想的な推定量は、その推定量の期待値が母数と同じであることが求められます。

代表的な標本統計量の1つである標本分散は、次のように定義されます。

$$
s^2 = \frac{1}{n-1} \sum_{k=1}^{n} (X_k - \overline{X})^2 \quad \tag{1}
$$

ここで $ n $ の代わりに $ n-1 $ で割って標本統計量を定義しているのは、こうすることで標本分散の期待値が母分散と一致するためです。

### 証明: 標本分散の期待値 = 母分散の値

標本のサイズが $ n $ であり、標本平均が $ \overline{X} $、標本分散が $ s^2 $ である標本を抽出したとします。このとき、母平均の値は $ m $、母分散の値は $ \sigma^2 $ です。

このとき、

$$
s^2 = \frac{1}{n-1} \sum_{k=1}^{n} (X_k - \overline{X})^2 \quad \tag{2}
$$

としたとき、$ E(s^2) = \sigma^2 $ であるかを確認しましょう。

$$
E[s^2] = \frac{1}{n-1} E\left[\sum_{k=1}^{n} (X_k - \overline{X})^2\right] \quad \tag{3}
$$

このとき、内部の総和についてのみ考えてみます。

$$
\sum_{k=1}^{n} (X_k - \overline{X})^2 = \sum_{k=1}^{n} \left((X_k - m) + (m - \overline{X})\right)^2 \quad \tag{4}
$$

$$
= \sum_{k=1}^{n} \left((X_k - m)^2 + 2(X_k - m)(m - \overline{X}) + (m - \overline{X})^2\right) \quad \tag{5}
$$

$$
= \sum_{k=1}^{n} (X_k - m)^2 + 2(\overline{X} - m)n(m - \overline{X}) + n(m - \overline{X})^2 \quad \tag{6}
$$

$$
= \sum_{k=1}^{n} (X_k - m)^2 - n(\overline{X} - m)^2 \quad \tag{7}
$$

したがって、

$$
E[s^2] = \frac{1}{n-1} E\left[\sum_{k=1}^{n} (X_k - \overline{X})^2\right] = \frac{1}{n-1} E\left[\sum_{k=1}^{n} (X_k - m)^2 - n(\overline{X} - m)^2\right] \quad \tag{8}
$$

と書くことができます。

このとき、

$$
E[(X_k - m)^2] = \sigma^2 \quad \tag{9}
$$

であり、

$$
E[(\overline{X} - m)^2] = V[\overline{X}] = \frac{\sigma^2}{n} \quad \tag{10}
$$

です。

したがって、

$$
E[s^2] = \frac{1}{n-1} E\left[\sum_{k=1}^{n} (X_k - \overline{X})^2\right] = \frac{1}{n-1} E\left[\sum_{k=1}^{n} (X_k - m)^2 - n(\overline{X} - m)^2\right] \quad \tag{11}
$$

となります。

このとき、期待値演算は線形演算であるため、

$$
= \frac{1}{n-1} \left[\sum_{k=1}^{n} E[(X_k - m)^2] - nE[(\overline{X} - m)^2]\right] \quad \tag{12}
$$

$$
= \frac{1}{n-1} (n\sigma^2 - n\frac{\sigma^2}{n}) = \sigma^2 \quad \tag{13}
$$

となります。

したがって、

$$
s^2 = \frac{1}{n-1} \sum_{k=1}^{n} (X_k - \overline{X})^2 \quad \tag{14}
$$

であってこそ、標本分散の期待値が母分散となることが証明されます。
