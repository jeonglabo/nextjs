##**失敗を祈る理論 - 帰無仮説と対立仮説**  


新しいことを学ぶ中で直面する難しさの一つは、どの分野でも避けられない「用語」です。用語に慣れて初めて、その分野の学習が進むこともよくあります。

統計学で初めて接する難しい用語の一つに「仮説」があります。代表的なのが帰無仮説と対立仮説ですが、漢字の用語であるために難しく感じられるだけでなく、英語で見ても「null hypothesis」「alternative hypothesis」となり、助けにはなりません。

筆者の場合、大学院に入ってから独学で統計学を学び始めましたが、これらの用語に慣れるまでに1年以上かかったように思います。

この記事では、これらの仮説に関連する用語について説明し、それがどのような意味を持ち、どのような理由でこのような概念が必要なのかを見ていきます。

### 帰無仮説と対立仮説

私たちは研究や開発を通じて新しい事実を発見し、それを証明したいと考えます。このとき、帰無仮説と対立仮説が使われますが、それぞれ「新しいことはない」という仮説と「新しいことがある」という仮説と考えることができます。

例えば、喫煙が脳疾患の発生率に影響を与えるかどうかを研究する場合、次のようになります。

- **帰無仮説**：喫煙の有無は脳血管疾患の発生に影響を与えない。  
- **対立仮説**：喫煙の有無は脳血管疾患の発生に影響を与える。

しかし、常識的な視点から考えると、疑問に思うことがあります。「新しいことがある」という対立仮説だけで十分なのに、なぜわざわざ「帰無仮説」という不必要に見える概念を追加するのでしょうか？

帰無仮説を導入する理由は、主に次の3つに集約できます。

1. 「真でないことを証明する方が、真であることを証明するよりもはるかに簡単だから」  
2. 「帰無仮説を『正しく』記述する方が、対立仮説を『正確に』記述するよりも失敗する可能性が少ないから」  
3. 「母数について知ることができないうえ、研究に主観性が入り込んではならないから」  

これらの理由から、伝統的な統計学では帰無仮説の検証に失敗することで、間接的に新しい仮説を確認しようとします。この方法は、帰謬法と同様と考えてもよいでしょう。

帰謬法とは、ある命題を間接的に証明する方法です。ある命題が真であることを直接証明するのが難しいため、その否定が真であると仮定し、証明を進めていく中で矛盾を発見し、それを通じてその命題を証明するのです。

つまり、調査や研究で何か変化があるという仮説を検証しようとするなら、逆に「仮説がない」と仮定して実験を行います。そして、変化がないという仮説に矛盾があることを見つけられれば、それを根拠に変化があることを間接的に証明できるのです。

帰無仮説を利用した検証方法は、「無罪推定の原則」と同じように比喩することができます。無罪推定の原則に従うと、容疑者や被告は有罪と判決が確定（帰無仮説が棄却された状態）するまでは無罪と推定されます（帰無仮説が棄却されていない状態）。このとき、有罪と判決するためには被告が実際に無罪だと仮定した場合には起こり得ない証拠や状況（統計学的に有意なレベル）が必要です。

### 帰無仮説・対立仮説についての注意点

帰無仮説や対立仮説に関して注意すべきポイントの一つは、帰無仮説を棄却できたからといって対立仮説を証明したことにはならない点です。つまり、帰無仮説の検証過程は、あくまで「（検証の）失敗」に重点を置いたプロセスなのです。この点は研究分野で非常に重要な問題であり、これについてはp値の説明をするときにさらに詳しく取り上げることにします。
