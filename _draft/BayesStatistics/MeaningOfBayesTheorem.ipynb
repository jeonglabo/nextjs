##**ベイズの定理の意味**  

### ベイズの定理の公式
まず、ベイズの定理の公式を確認しましょう。ベイズの定理の公式は、以下の式 (1) のようになります。

$$
P(H | E) = \frac{P(E | H) P(H)}{P(E)} \quad (1)
\]

式 (1) には4つの確率値が書かれており、見た目も似ているため、ぱっと見ただけでは意味を理解するのが難しいです。

4つの確率値のうち、$P(H)$ と $P(H | E)$ はそれぞれ事前確率と事後確率と呼ばれ、ベイズの定理は基本的に事前確率と事後確率の関係を示す定理です。

では、事前確率と事後確率の意味を把握することで、ベイズの定理が示す内容とその意義を理解することができるでしょう。

### ベイズの定理の意味と意義
結論から言えば、ベイズの定理は新しい情報に基づいて、ある事象が発生したという主張に対する信頼度を更新する方法（新しい情報をもとに信念を更新する方法）です。

### ベイズの定理が理解しづらい理由
ベイズの定理を理解する際に最初に整理すべき概念は「確率」に関する見方です。

ベイズの定理の意味を理解しにくい理由の一つは、高校までの確率論では「伝統的な視点」で確率を定義し、理解してきたためです。

ここでは、確率という言葉を「主張に対する信頼度」として考えてみましょう。

この視点は、確率に対するベイジアン主義（Bayesianism）の観点と見ることができます。対照的に、伝統的な確率観は頻度主義（frequentism）と見なされます。

たとえば、「コインの表が出る確率が50%」と言った場合、頻度主義者は100回コインを投げると50回は表が出ると解釈し、ベイジアン主義者は「コインの表が出るという主張の信頼度が50%」と見なします。

### 用語の整理
また、式 (1) の公式で $E$ と $H$ が何を意味するのかを見てみましょう。

式 (1) で $H$ は Hypothesis（仮説）の略で、「ある事象が発生したという主張」を意味します。

さらに、式 (1) の $E$ は Evidence（証拠）の略で、「新しい情報」を意味します。

したがって、$P(H)$ は「ある事象が発生したという主張に関する信頼度」を意味し、$P(H | E)$ は「新しい情報を得た後に更新された信頼度」を意味します。

そして、$P(H)$ と $P(H | E)$ はそれぞれ事前確率と事後確率と呼ばれます。事前（事前）、事後（事後）という言葉を考えれば、何らかの事（ここでは証拠）を観測する前後の自分の主張に関する信頼度と理解できます。

**図1. ベイズの定理は事前確率と事後確率の関係を説明する定理です。**

### 確率論のパラダイムシフト: 演繹的推論から帰納的推論へ
ベイズの定理が画期的である理由は、統計学の基本的なパラダイムを変えたからです。

従来の統計学は、前述のように「頻度主義」の観点に基づいて構成されており、すべて演繹的な思考に基づいています。

たとえば、全体の事象集合について把握して特定の事象が起こる確率を計算する、あるいは集団の統計的分布を仮定して差の有意性を検定する、といった方法が考えられます。

つまり、従来の統計学では、厳密に確率空間を定義したり、集団（母集団または標本集団）の分布を定義し、その後に計算によって導かれる結果を受け入れるパラダイムを採用しています。

一方、ベイジアンの観点の統計学では、事前確率のような経験に基づく先験的な、あるいは不確実性を含む数値に基づいており、それに追加情報を基にして事前確率を更新します。

このような方法は帰納的推論の方法であり、ベイジアンの視点の確率論・統計学は、追加の証拠を得ることで真理に近づくことができるという哲学を含んでおり、

確率論のパラダイムに大きな変革をもたらしたと言えます。

### 問題解決を通じたベイズの定理の理解
簡単な例題を解きながら、ベイズの定理が新しい情報に基づいてある事象が発生したという主張に対する信頼度を更新していく方法であることを理解しましょう。
続きの翻訳を以下に示します。





### 例題 1
ある病気Aの発病率は0.1%であると知られています。この病気が実際に存在するときに病気があると診断する確率（感度）は99%、病気がないときに病気がないと正しく診断する確率（特異度）は98%とします。

もしある人がこの病気にかかっていると診断されたとき、その人が本当に病気にかかっている確率はどのくらいでしょうか？

### 解答
このようなベイズの定理を利用した問題をうまく解決するためには、先に述べた仮説（Hypothesis）と証拠（Evidence）を正しく定義する必要があります。

この問題では、仮説と証拠は次のように定義できます。

- **Hypothesis**: 真である、つまり実際に病気がある。
- **Evidence**: 陽性である、つまり病気があると診断された。

したがって、基本的に病気Aの発病率が0.1%なので、任意の人がこの病気にかかる確率は \(P(H)\) で表され、\(P(H) = 0.001\) です。

また、問題で与えられた感度と特異度は、それぞれ真陽性率と真陰性率に対応しており、それぞれ \(P(E | H) = 0.99\)、\(P(E^c | H^c) = 0.98\) です。

これを図で示すと、次の図2のようになります。

**図2. 与えられた確率値を長方形の相対的な面積で視覚化したもの**

式 (1) をもう少し整理して書くと、以下の式 (2) のようになります。

$$
P(H | E) = \frac{P(E | H) P(H)}{P(E)} = \frac{P(E | H) P(H)}{P(E | H) P(H) + P(E | H^c) P(H^c)} \quad (2)
$$

したがって、求めたい $P(H | E)$ を計算すると、次のようになります。
$$
P(H | E) = \frac{0.001 \times 0.99}{0.001 \times 0.99 + 0.02 \times 0.999} = 0.047 \quad \text{（小数点以下3桁まで丸め）} \quad (3)
$$

### 例題 2
例題1で一度陽性判定を受けた人が、二度目の検診で再び陽性判定を受けたとき、この人が本当に病気にかかっている確率はどのくらいでしょうか？

### 解答
例題2のような問題を出す理由は、ベイズの定理が信頼度を更新する方法であると述べたからです。例題2のような状況では、仮説を次のように変えて考えることができます。

- **Hypothesis**: 病気があると一度診断されたことが事実である。

つまり、例題1で計算された事後確率の値が、例題2では事前確率として使われ、再度更新された事後確率を計算することになります。

**図3. 例題1で計算した事後確率が、例題2で事前確率として再度使用され（赤い矢印）、これを用いてもう一度更新された事後確率を計算することができます。図は、与えられた確率値を長方形の相対的な面積で視覚化したもの。**

したがって、求めたい $P(H | E)$ を計算すると、次のようになります。

$$
P(H | E) = \frac{0.047 \times 0.99}{0.047 \times 0.99 + 0.02 \times 0.953} = 0.709 \quad \text{（小数点以下3桁まで丸め）} \quad (4)
$$

つまり、例題1で計算して得た事前確率に基づいてもう一度確率（つまり、病気にかかっているという事実に対する信頼度）を更新すると、約70%の信頼度でこの人は病気にかかっていると言うことができるでしょう。
