### 前提条件

### ベクトルの射影

中高等学校の数学で、ベクトルの射影について学んだことがあると思います。

射影の概念が成り立つためには、2つのベクトルが必要です。

例えば、以下の図ではベクトル $ \mathbf{a} $ とベクトル $ \mathbf{b} $ の関係を示しています。

---

#### 図1: 2つのベクトルのうち、1つのベクトルに対してもう1つのベクトルが向かう方向の成分は、余弦値を使って表現できる。

ベクトル $ \mathbf{b} $ をベクトル $ \mathbf{a} $ を使って表現する場合、ベクトル $ \mathbf{b} $ の $ \mathbf{a} $ 方向の成分は、次のように表すことができます。

$$
\text{comp}_{\mathbf{a}} \mathbf{b} = |\mathbf{b}| \cos \theta
\tag{1}
$$

### グラム・シュミットのプロセス

グラム・シュミットのプロセスの目的は以下の通りです。

「線形独立なベクトルが与えられたとき、それらを適切に変形して直交基底に変換する」

---

#### 図3: グラム・シュミットのプロセスは、与えられた2つのベクトルを直交する2つのベクトルに変換するプロセス。

直交するベクトルセットを得ると、特に直交しない基底に比べて利便性が向上します。

---

### グラム・シュミットのプロセスの数式的表現

ベクトル空間 $ V $ 上に与えられた一次独立なベクトル $ \mathbf{a}_1, \mathbf{a}_2, \dots, \mathbf{a}_k $ を用いて、次のように直交基底を求めることができます。

$$
\mathbf{u}_1 = \mathbf{a}_1
\tag{9}
$$
$$
\mathbf{u}_2 = \mathbf{a}_2 - \text{proj}_{\mathbf{u}_1} (\mathbf{a}_2)
\tag{10}
$$
$$
\mathbf{u}_3 = \mathbf{a}_3 - \text{proj}_{\mathbf{u}_1} (\mathbf{a}_3) - \text{proj}_{\mathbf{u}_2} (\mathbf{a}_3)
\tag{11}
$$
$$
\vdots
$$
$$
\mathbf{u}_k = \mathbf{a}_k - \sum_{j=1}^{k-1} \text{proj}_{\mathbf{u}_j} (\mathbf{a}_k)
\tag{12}
$$

---

### QR分解

QR分解は、グラム・シュミットのプロセスを用いて正規直交基底ベクトルを行列に分解するプロセスです。

$$
A = QR
\tag{14}
$$

---

### 例題

次の行列 $ A $ をQR分解しなさい。

$$
A = \begin{bmatrix} 1 & 1 & 0 \\ 1 & 0 & 1 \\ 0 & 1 & 1 \end{bmatrix}
\tag{17}
$$

行列 $ A $ の各列ベクトルを使って、グラム・シュミットのプロセスを適用し、QR分解を実行します。

---

最終的に、QR分解の結果は次のようになります。

$$
A = QR = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{6}} & -\frac{1}{\sqrt{3}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{6}} & \frac{1}{\sqrt{3}} \\ 0 & \frac{2}{\sqrt{6}} & \frac{1}{\sqrt{3}} \end{bmatrix}
\begin{bmatrix} \frac{2}{\sqrt{2}} & \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ 0 & \frac{3}{\sqrt{6}} & \frac{1}{\sqrt{6}} \\ 0 & 0 & \frac{2}{\sqrt{3}} \end{bmatrix}
\tag{34}
$$
