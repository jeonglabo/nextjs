### 回転行列の固有値と固有ベクトル

$$
A = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix}
\tag{1}
$$

例えば、90度反時計回りに回転する行列を適用した線形変換の結果は以下のようになります。

$$
\begin{bmatrix} \cos(\frac{\pi}{2}) & -\sin(\frac{\pi}{2}) \\ \sin(\frac{\pi}{2}) & \cos(\frac{\pi}{2}) \end{bmatrix}
\tag{2}
$$

### 固有値と固有ベクトルの意味

前の投稿「固有値と固有ベクトルの意味」で述べた核心は以下の通りです。

「ベクトル $ x $ に何らかの線形変換 $ A $ を行った際、ベクトルの大きさだけが変わり、元のベクトルと平行なベクトル $ x $ とは何でしょうか？」  
「その大きさはどのくらい変わりますか？」

---

### 固有値と固有ベクトルの計算

#### 固有値の計算

$$
A \vec{x} = \lambda \vec{x}
\tag{3}
$$

$$
\begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix} \vec{x} = \lambda \vec{x}
\tag{4}
$$

ここで $ \vec{x} $ は、単位行列 $ I $ を用いて次のように分解して考えられます。

$$
I = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
\tag{5}
$$

この式を変形すると、

$$
(A - \lambda I) \vec{x} = \begin{bmatrix} \cos(\theta) - \lambda & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) - \lambda \end{bmatrix} \vec{x} = \vec{0}
\tag{6}
$$

ここで、$ \vec{x} $ が零ベクトルでないためには、行列 $ (A - \lambda I) $ が逆行列を持たない必要があり、行列式が0である必要があります。

$$
\det(A - \lambda I) = (\cos(\theta) - \lambda)^2 + \sin^2(\theta) = 0
\tag{7}
$$

$$
\cos^2(\theta) - 2\lambda \cos(\theta) + \lambda^2 + \sin^2(\theta) = 0
\tag{8}
$$

ここで、$ \cos^2(\theta) + \sin^2(\theta) = 1 $ なので、

$$
\lambda^2 - 2\lambda \cos(\theta) + 1 = 0
\tag{9}
$$

この2次方程式を解くと、固有値 $ \lambda $ は次のようになります。

$$
\lambda = \cos(\theta) \pm \sqrt{\cos^2(\theta) - 1}
\tag{10}
$$

$$
\lambda = \cos(\theta) \pm i \sin(\theta)
\tag{11}
$$

ここで $ i = \sqrt{-1} $ です。

---

#### 固有ベクトルの計算

1. $ \lambda = \cos(\theta) + i \sin(\theta) $ の場合の固有ベクトル

次の方程式を満たす固有ベクトル $ \vec{x} $ を求めます。

$$
\begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix} \vec{x} = (\cos(\theta) + i \sin(\theta)) \vec{x}
\tag{12}
$$

これを解くと、

$$
\vec{x} = \begin{bmatrix} i \\ 1 \end{bmatrix}
\tag{13}
$$

---

2. $ \lambda = \cos(\theta) - i \sin(\theta) $ の場合の固有ベクトル

同様にして、この場合の固有ベクトルは次のようになります。

$$
\vec{x} = \begin{bmatrix} -i \\ 1 \end{bmatrix}
\tag{14}
$$

---

### 固有値と固有ベクトルの別の視点

「固有ベクトルに対する線形変換は、固有値の分だけ拡大または縮小される。」という視点で捉えることができます。

---

### 複素固有値とベクトルの回転

複素数の掛け算は、ベクトルの回転を表します。したがって、複素固有値はベクトルの長さを変えるのではなく、「複素数の掛け算を通じてベクトルを回転させる」という意味を持ちます。
