### 疑似逆行列の幾何学的意味  
2020年11月11日

---

※ 疑似逆行列の正式な名称は「ムーア・ペンローズ擬似逆行列」ですが、一般的に多く使用されている「疑似逆行列（pseudo inverse）」を使用します。

※ 疑似逆行列は複素数体の範囲でも定義可能ですが、今回は実数体の範囲で説明します（視覚化目的および計算の混乱を避けるため）。

※ 疑似逆行列は基本的に線形回帰における線形代数学的アプローチと完全に同じアプローチを取ります。

---

### 前提条件

この投稿を理解するためには、以下の内容を知っているとより良いです。

- 行ベクトルの意味とベクトルの内積：特に双対空間の概念
- 特異値分解（SVD）

---

### 疑似逆行列の定義

まず、最もシンプルに疑似逆行列の定義から確認します。

疑似逆行列は、任意の行列 $ A \in \mathbb{R}^{m \times n} $ に対して次のように定義されます。

1. $ m > n $ で、全ての列ベクトルが線形独立の場合

$$
A^+ = (A^T A)^{-1} A^T
\tag{1}
$$

このとき、$ A^T A $ は可逆行列である必要があります。この場合、$ A^+ A = I $ が成立するため、$ A^+ $ は左逆行列（left inverse）と呼ばれます。

2. $ m < n $ で、全ての行ベクトルが線形独立の場合

$$
A^+ = A^T (A A^T)^{-1}
\tag{2}
$$

このとき、$ A A^T $ は可逆行列である必要があります。この場合、$ A A^+ = I $ が成立するため、$ A^+ $ は右逆行列（right inverse）と呼ばれます。

---

### 疑似逆行列の数式的意味

次に、疑似逆行列が持つ基本的な数式的意味について考えてみます。

例えば、次のような連立方程式が与えられたとします。

$$
A \mathbf{x} = \mathbf{b}
\tag{3}
$$

$$
\Rightarrow \begin{bmatrix} -1 & 1 \\ 0 & 1 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \\ 3 \end{bmatrix}
\tag{4}
$$

---

### 疑似逆行列の幾何学的意味

連立方程式は、2つ以上の未知数を含む方程式の組を指します。通常の連立方程式の形は以下のように表されます。

$$
\begin{cases}
ax + by = p \\
cx + dy = q
\end{cases}
\tag{10}
$$

しかし、今回は未知数の数よりデータの方が多い場合に対して適切な解を見つける過程を通じて、疑似逆行列を考えていきます。

---

### 最適な解の探求

例えば、次のような3つのデータポイントが与えられたとします。

$$
(-1, 0), (0, 1), (0, 3)
\tag{11}
$$

この3点に対して $ f(x) = mx + b $ のようなモデルを用いて、次の連立方程式を立てることができます。

$$
f(-1) = -m + b = 0
\tag{12}
$$
$$
f(0) = b = 1
\tag{13}
$$
$$
f(0) = b = 3
\tag{14}
$$

---

### 最適な解を求めるための線形代数的アプローチ

幾何学的には、3つのデータポイントを通過する直線を見つける問題に対応します。しかし、どのように線を引いても3点を同時に通過する直線は存在しません。よって、この問題には解が存在しません。

---

### 疑似逆行列の最終的な導出

このように、疑似逆行列は元の方程式の解を必ずしも与えませんが、与えられたデータに最も近い解を提供するものとして使用されます。
