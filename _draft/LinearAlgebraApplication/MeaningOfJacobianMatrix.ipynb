### ヤコビ行列の幾何学的な意味

**※** Jacobian matrixの正式な日本語名は「ヤコビ行列」ですが、一般的に多く使用されている「ヤコビアン行列」として呼びます。  
**※** この投稿では列ベクトル（column vector）規約に従います。  
**※** 本記事の内容は主にGrant SandersonのKhan Academyのヤコビアン講義から引用しています。  

#### 前提条件
ヤコビアン行列の幾何学的な意味を理解するために、次の内容を知っていると良いです。

- 行列と線形変換
- 行列式の幾何学的な意味
- 重積分の意味

### ヤコビアン行列の定義

まずはヤコビアン行列の定義から始めましょう。

$ n $次元ベクトル $ \mathbf{x} \in \mathbb{R}^n $ を入力として、$ m $次元ベクトル $ \mathbf{f}(\mathbf{x}) \in \mathbb{R}^m $ を出力するベクトル関数 $ \mathbf{f} : \mathbb{R}^n \to \mathbb{R}^m $ があると仮定します。

このとき、この関数の1次偏微分が $ \mathbb{R}^n $ の実数ベクトル空間で存在する場合、ヤコビアンは次のような $ m \times n $ 行列として定義できます。

$$
\mathbf{J} =
\begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{bmatrix}
\tag{1}
$$

ヤコビアン行列を見ると、この行列のすべての要素が1次微分係数で構成されていることがわかります。また、ヤコビアン行列は微小変化に関する線形変換という点でも重要な意味を持ちます。

実際、ヤコビアンが示すのは、微小領域で「非線形変換」を「線形変換で近似」したものであるということです。この点についてさらに理解を深めていきましょう。

### 多変数関数のチェーンルール

ヤコビアン行列を理解する上で、重要なチェーンルールについて簡単に説明します。

一般的に多変数関数とは、2つ以上の入力を持つ関数を指しますが、後の例ではすべて2次元平面に表示できる二変数関数を使用してチェーンルールを説明します。

二変数関数 $ z = f(x, y) $ において、$ x = g(t) $、$ y = h(t) $ であり、$ f(x, y) $、$ g(t) $、$ h(t) $ がすべて微分可能な関数であるとすると、

$$
\frac{dz}{dt} = \frac{\partial z}{\partial x} \frac{dx}{dt} + \frac{\partial z}{\partial y} \frac{dy}{dt}
\tag{2}
$$

この式を少し変形すると次のように書けます。

$$
dz = \frac{\partial z}{\partial x} dx + \frac{\partial z}{\partial y} dy
\tag{3}
$$

この式は、チェーンルールを用いて多変数関数で各変数の変化が関数全体に与える影響を表しています。この概念はヤコビアン行列の構成要素と密接に関連しており、ヤコビアン行列は各変数の微小な変化が関数全体に与える変換効果を示すツールとして使用されます。

ヤコビアン行列は、与えられた関数の微分可能性や局所的な線形近似の重要性を理解する上で重要な役割を果たします。このため、さまざまな分野で非線形変換を扱う際の重要な数学的ツールとして活用されています。


### ヤコビアン（Jacobian）行列の幾何学的意味

#### 非線形変換

**行列と線形変換**では、線形代数学の基本定理を説明しました。簡単に言えば、行列は線形変換だということです。

幾何学的に見ると、線形変換は以下の特徴を持つ変換を意味します：

1. 変換後も原点の位置が変わらない。
2. 変換後も格子の形が直線の形を維持する。
3. 格子間の間隔が均等に保たれる。

代表的な線形変換であるシアー(shear)行列の線形変換のアニメーションを例として見てみましょう。

#### 線形変換の例：シアー行列

$$
\begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
\tag{4}
$$

上記のシアー行列の変換前と変換後の格子の形を見ると、線形変換の幾何学的特徴の3つを満たしていることがわかります。

一方、**非線形変換**は、簡単に言えば、線形変換ではない変換を指します。以下の非線形変換の例を見て、幾何学的に線形変換の特徴を持たないことを視覚的に確認してみてください。

#### 非線形変換の例

$$
f(x, y) =
\begin{bmatrix}
x + \sin\left(\frac{y}{2}\right) \\
y + \sin\left(\frac{x}{2}\right)
\end{bmatrix}
\tag{5}
$$

#### 非線形変換を局所的に観察すると？

前述したヤコビアンの定義では、ヤコビアンが指すものは微小領域での「非線形変換」を「線形変換で近似」したものだと述べました。それでは、本当に非線形変換を微小領域で見ると、線形変換として十分に近似できるのでしょうか？

次の視覚的な例を確認してみましょう。左のグラフにある白いボックスのシーンを拡大したものが右のグラフに描かれています。

この非線形変換の局所的な観察に関する最後のシーンは、以下の図1に示されています。

![図1](https://example.com)  
**図1. 非線形変換を局所的に観察すると、線形変換で近似できることがわかる。**

図1に見られるように、非線形変換を局所的に観察すると、変換後も格子の形が直線に近く、格子間の間隔も均等に保たれていることがわかります。

#### 原点の位置の問題を解決するには？

初めに述べた線形変換の幾何学的特徴の中で、原点の位置が変わらないという問題をどのように解決するのでしょうか？

図2に示されるように、変換したい点 $(x_0, y_0)$ を原点として考え、その点での変換を近似する行列を得れば、それが非線形変換を線形変換として近似した行列である「ヤコビアン行列」となります。

このように、ヤコビアン行列を使うことで、非線形変換も局所的には線形変換として近似でき、この行列はその点の周辺での変化を説明する重要な道具となります。


図 2. ヤコビ行列は非線形変換を線形変換で近似した結果です。
ヤコビ行列は、非線形変換を特定の点での線形変換として近似したものです。この行列は変換の勾配（微分係数）を表し、非線形関数が局所的に線形のように振る舞う方法を示します。つまり、複雑な非線形変換を単純な線形変換で近似できる強力なツールです。

出典: Math Stack Exchangeでは、ヤコビ行列に関する詳細な説明と例を見ることができます。このリンクでは、ヤコビ行列の定義に加え、与えられた方程式を用いてヤコビ行列を構成する方法について具体的な例が示されています。これにより、ヤコビ行列が非線形変換をどのように近似するかについて視覚的な理解を深めることができます。

### ヤコビ行列の導出

ヤコビ行列が非線形変換を局所的に線形変換として近似するものであることを理解したところで、実際にヤコビ行列を導出してみましょう。

次の図のように、非線形変換の結果を線形変換として近似する場合、$(u, v)$ 座標系から $(x, y)$ 座標系に変換されると考えることができます。

#### 図 3. 非線形変換前後の微小変位の関係

すると、ある線形変換 $J$ によって $du$ と $dv$ が $dx$ と $dy$ に変換されると考えられます。これを数式で表すと次のようになります。

$$
\begin{bmatrix}
dx \\
dy
\end{bmatrix}
= J
\begin{bmatrix}
du \\
dv
\end{bmatrix}
=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\begin{bmatrix}
du \\
dv
\end{bmatrix}
\tag{6}
$$

この式を展開すると次のようになります。

$$
dx = a \times du + b \times dv \tag{7}
$$

$$
dy = c \times du + d \times dv \tag{8}
$$

この2つの式は、チェーンルールを使用して、局所的な非線形変換の前後の関係を表しており、これによりヤコビ行列は次のように考えることができます。

$$
J =
\begin{bmatrix}
\frac{\partial x}{\partial u} & \frac{\partial x}{\partial v} \\
\frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}
\end{bmatrix}
\tag{9}
$$

### ヤコビ行列の行列式

行列式の幾何学的な意味において、行列式は線形変換時に単位面積がどれだけ増加するかを表します。図4に示すように、線形変換前の四角形の面積が1であれば、線形変換後の面積は行列式の値である $ad - bc$ だけ変化します。

#### 図 4. 行列式の幾何学的な意味：行列式は線形変換時の単位面積の変化を示します。

したがって、ヤコビ行列の行列式は、元の座標系から変換された座標系に変換される際の面積の変化率を示します。図5に示すように、$(u, v)$ 座標系から $(x, y)$ 座標系に変換される場合、$dx \times dy$ と $du \times dv$ の関係は次のようになります。

$$
dx \times dy = |J| (du \times dv) \tag{10}
$$

ここで、$|J|$ はヤコビ行列の行列式を意味し、これは線形変換時の単位面積がどれだけ変化するかを示します。

#### 図 5. 行列式の幾何学的な意味：行列式は線形変換時の単位面積の変化を示します。


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

# 設定
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# 左のグラフ: 単位正方形
ax[0].set_xlim(-0.5, 2)
ax[0].set_ylim(-0.5, 2)
ax[0].grid(True)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

# 単位正方形の頂点
unit_square = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])

# 単位正方形の描画
rect = Polygon(unit_square, closed=True, fill=True, color='green', alpha=0.3, edgecolor='r')
ax[0].add_patch(rect)
ax[0].plot(*unit_square.T, 'ko-')
ax[0].text(1, 0, '(1, 0)', fontsize=12, ha='right')
ax[0].text(0, 1, '(0, 1)', fontsize=12, va='bottom')
ax[0].text(0, 0, 'O', fontsize=12, ha='right', va='top')

# タイトル
ax[0].set_title('Unit Square')

# 右のグラフ: 変換後の平行四辺形
ax[1].set_xlim(-1, 5)
ax[1].set_ylim(-1, 5)
ax[1].grid(True)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

# 行列の定義 [a b; c d]
a, b, c, d = 2, 1, 1, 3

# 変換後の頂点の計算
O = np.array([0, 0])               # 原点
R = np.array([a, 0])               # (a, 0)
P = np.array([0, d])               # (0, d)
Q = np.array([a, d])               # (a, d)

# 平行四辺形の頂点
parallelogram = np.array([O, R, Q, P])

# 変換後の平行四辺形の描画
trans_rect = Polygon(parallelogram, closed=True, fill=True, color='orange', alpha=0.3, edgecolor='g')
ax[1].add_patch(trans_rect)

# 重要な点の描画
ax[1].plot(*O, 'ko')
ax[1].plot(*R, 'ko')
ax[1].plot(*P, 'ko')
ax[1].plot(*Q, 'ko')

# 各点のラベル
ax[1].text(*O, 'O', fontsize=12, ha='right', va='top')
ax[1].text(*R, f'R({a}, 0)', fontsize=12, ha='right', va='top')
ax[1].text(*P, f'P(0, {d})', fontsize=12, ha='left', va='top')
ax[1].text(*Q, f'Q({a}, {d})', fontsize=12, ha='left', va='top')

# タイトル
ax[1].set_title('Transformed Parallelogram')

# 面積のテキスト表示
det = a * d - b * c
ax[1].text(2, 4, f'Area of Parallelogram: ad - bc = {det}', fontsize=12, ha='center')

plt.suptitle('Geometric Meaning of Determinant')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, FancyArrowPatch

# 図の設定
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# 左のグラフ: 変換前の微小領域
ax[0].set_xlim(-1, 3)
ax[0].set_ylim(-1, 3)
ax[0].grid(True)
ax[0].set_xlabel('u')
ax[0].set_ylabel('v')

# 微小領域（四角形）の定義
rect = Polygon([[0, 0], [2, 0], [2, 2], [0, 2]], closed=True, color='lightblue', alpha=0.6, edgecolor='k')
ax[0].add_patch(rect)
ax[0].text(1, 2.1, 'Area: du × dv', fontsize=12, ha='center')
ax[0].text(0, 2, 'dv', fontsize=12, ha='right', va='center')
ax[0].text(2, 0, 'du', fontsize=12, ha='center', va='top')

# タイトル
ax[0].set_title('Before Nonlinear Transformation')

# 右のグラフ: 変換後の領域
ax[1].set_xlim(-1, 3)
ax[1].set_ylim(-1, 3)
ax[1].grid(True)
ax[1].set_xlabel('u')
ax[1].set_ylabel('v')

# 変換後の平行四辺形の定義
parallelogram = Polygon([[0, 0], [2, 0.5], [2.5, 2.5], [0.5, 2]], closed=True, color='salmon', alpha=0.6, edgecolor='k')
ax[1].add_patch(parallelogram)
ax[1].text(1.5, 2.6, 'Area: dx × dy', fontsize=12, ha='center')
ax[1].text(0.5, 2, 'dy', fontsize=12, ha='right', va='center')
ax[1].text(2.5, 0.5, 'dx', fontsize=12, ha='center', va='top')

# タイトル
ax[1].set_title('After Nonlinear Transformation')

# 中央の矢印の描画
arrow = FancyArrowPatch((2.5, 1.5), (0.5, 1.5), mutation_scale=20, color='green')
fig.patches.append(arrow)
ax[0].text(3, 1.5, 'Nonlinear Transformation', fontsize=14, ha='center', va='center', color='green')

# 図全体のタイトル
plt.suptitle('Geometric Meaning of Nonlinear Transformation')
plt.tight_layout()
plt.show()
```

##ヤコビアン行列式の使用例

今回はヤコビアンを利用して円の面積を求める例を紹介します。

円の面積を求める方法はさまざまですが、極座標 $ r, \theta $ で面積を求める方法も一つの手段です。では、ヤコビアンを使って $ r, \theta $ 座標系で円の面積を求めることはどういう意味でしょうか？

これは、横軸が $ r $、縦軸が $ \theta $ の座標系で何かを計算した後（ここでは $ r, \theta $ 座標系での面積）に、横軸が $ x $、縦軸が $ y $ の座標系に変換することを意味します。このとき、ヤコビアン行列式の役割は、座標変換時に必要な面積の補正値と言えます。

これを理解するために、以下の2つのアニメーションを観察してください。

アニメーションの最初のシーンでは、$ r, \theta $ 座標系にあった点を観察でき、後半ではその点が $ x, y $ 座標系に移動しているのがわかります。この過程で、左のアニメーションでは円の面積が隙間だらけに見えますが、右のアニメーションでは円の面積がある程度きっちり埋まっているのがわかります。このとき、適切に面積を埋める補正値がヤコビアンの行列式の値です。

$ r, \theta $ 座標系から $ x, y $ 座標系への変換式は以下の通りです。

$$
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
r \cos(\theta) \\
r \sin(\theta)
\end{bmatrix}
\tag{11}
$$

この変換式からヤコビアン行列を計算すると、
$$
J= \begin{bmatrix}
\frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} \\
\frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta}
\end{bmatrix}
\tag{12}
$$

$$
=
\begin{bmatrix}
\cos(\theta) & -r\sin(\theta) \\
\sin(\theta) & r\cos(\theta)
\end{bmatrix} \tag{13}
$$

となり、ヤコビアンの行列式を計算すると次のようになります。

$$
|J| = r\cos^2(\theta) + r\sin^2(\theta) = r \tag{14}
$$

では、このヤコビアンの行列式を使って、$ r, \theta $ 座標系で半径3の円の面積を求めてみましょう。

$$
\int_{r=0}^{3} \int_{\theta=0}^{2\pi} dxdy \tag{15}
$$

$$
= \int_{r=0}^{3} \int_{\theta=0}^{2\pi} |J| dr d\theta \tag{16}
$$

$$
= \int_{r=0}^{3} \int_{\theta=0}^{2\pi} r dr d\theta \tag{17}
$$

$$
= \int_{r=0}^{3} r \left[ \theta \right]_{\theta=0}^{2\pi} dr \tag{18}
$$

$$
= 2\pi \int_{r=0}^{3} r dr \tag{19}
$$

$$
= 2\pi \cdot \frac{1}{2} r^2 \bigg|_{r=0}^{3} = 2\pi \cdot \frac{1}{2} \cdot 9 = 9\pi \tag{20}
$$

このようにして、ヤコビアンを利用して円の面積を求めることができます。
