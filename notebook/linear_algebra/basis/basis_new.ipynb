# 標準規定から新しい規定への変換

このページでは、標準規定から新しい規定への変換を視覚的にわかりやすくしています。

ぜひプログラム内の値を変更してみてください。

```python
import numpy as np
import matplotlib.pyplot as plt

# 基底行列と座標変換
basis_matrix = np.array([[1, 1], [1, -1]])  # 新しい基底 c1, c2
original_basis = np.eye(2)  # 標準基底 e1, e2
point = np.array([4, 2])  # 点 A の座標（標準基底）

# 点を新しい基底で表現
inverse_basis = np.linalg.inv(basis_matrix)
new_coords = np.dot(inverse_basis, point)

# グラフの作成
fig, ax = plt.subplots(figsize=(8, 8))

# 標準基底の描画（青色）とラベル
ax.quiver(0, 0, original_basis[0, 0], original_basis[1, 0], angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(original_basis[0, 0] * 1.1, original_basis[1, 0] * 1.1, r'$e_1$', color='blue', fontsize=12)
ax.quiver(0, 0, original_basis[0, 1], original_basis[1, 1], angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(original_basis[0, 1] * 1.1, original_basis[1, 1] * 1.1, r'$e_2$', color='blue', fontsize=12)

# 新しい基底の描画（緑色）とラベル
ax.quiver(0, 0, basis_matrix[0, 0], basis_matrix[1, 0], angles='xy', scale_units='xy', scale=1, color='green')
ax.text(basis_matrix[0, 0] * 1.1, basis_matrix[1, 0] * 1.1, r'$c_1$', color='green', fontsize=12)
ax.quiver(0, 0, basis_matrix[0, 1], basis_matrix[1, 1], angles='xy', scale_units='xy', scale=1, color='green')
ax.text(basis_matrix[0, 1] * 1.1, basis_matrix[1, 1] * 1.1, r'$c_2$', color='green', fontsize=12)

# 点 A を指す矢印とラベル
ax.quiver(0, 0, point[0], point[1], angles='xy', scale_units='xy', scale=1, color='red')
ax.text(point[0] * 1.05, point[1] * 1.05, r'$A$', color='red', fontsize=12)

# グリッド線の描画（新しい基底に沿った補助線）
x_range = np.arange(-5, 6)
y_range = np.arange(-5, 6)
for i in x_range:
    for j in y_range:
        grid_point = i * basis_matrix[:, 0] + j * basis_matrix[:, 1]
        ax.scatter(grid_point[0], grid_point[1], color='gray', alpha=0.6, s=10)  # グリッド点

for i in x_range:
    start = i * basis_matrix[:, 0] - 5 * basis_matrix[:, 1]
    end = i * basis_matrix[:, 0] + 5 * basis_matrix[:, 1]
    ax.plot([start[0], end[0]], [start[1], end[1]], color='gray', linestyle='--', alpha=0.5)

for j in y_range:
    start = j * basis_matrix[:, 1] - 5 * basis_matrix[:, 0]
    end = j * basis_matrix[:, 1] + 5 * basis_matrix[:, 0]
    ax.plot([start[0], end[0]], [start[1], end[1]], color='gray', linestyle='--', alpha=0.5)

# 軸の設定
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_aspect('equal', adjustable='box')

# 標準基底の目盛りを 1 間隔で設定
ax.set_xticks(np.arange(-6, 7, 1))
ax.set_yticks(np.arange(-6, 7, 1))
ax.set_xlabel('x')
ax.set_ylabel('y')

# タイトル
plt.title('Standard Basis to New Basis')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)

# グラフの表示
plt.show()
```
