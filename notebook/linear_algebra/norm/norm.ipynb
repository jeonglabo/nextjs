# ベクトルのノルム

ここではベクトルのノルムを視覚的に表示するプログラムが書かれています。

ぜひプログラム内の値を変更してみてください。

```python
import numpy as np
import matplotlib.pyplot as plt

# ベクトルを定義
vector = np.array([3, 4])  # ベクトル (3, 4)

# ノルムを計算
norm = np.linalg.norm(vector)

# 図を作成
fig, ax = plt.subplots(figsize=(6, 6))

# ベクトルを描画
ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\vec{v}$')

# ベクトルの成分を描画（点線を太く）
ax.plot([0, vector[0]], [0, 0], 'r--', linewidth=2, label=r'$x=3$')
ax.plot([vector[0], vector[0]], [0, vector[1]], 'g--', linewidth=2, label=r'$y=4$')

# ノルム（斜辺）を描画
ax.plot([0, vector[0]], [0, vector[1]], 'b', linewidth=2, label=r'$\sqrt{x^2+y^2}=5$')

# xの長さを矢印で強調（点線から少し下げる）
ax.annotate('', xy=(3, -0.5), xytext=(0, -0.5),
            arrowprops=dict(facecolor='red', arrowstyle='<|-|>', lw=1.5))
ax.text(1.5, -0.9, 'x=3', color='red', fontsize=12)

# yの長さを矢印で強調（点線から少し右にずらす）
ax.annotate('', xy=(3.5, 4), xytext=(3.5, 0),
            arrowprops=dict(facecolor='green', arrowstyle='<|-|>', lw=1.5))
ax.text(3.7, 2, 'y=4', color='green', fontsize=12)

# 斜辺の長さをテキストで表示（ベクトルから少し外す）
ax.text(0.4,2.5, r'$\sqrt{x^2+y^2}=5$', color='blue', fontsize=12)

# 直角記号を追加
rect_x = [vector[0] - 0.3, vector[0] - 0.3, vector[0]]
rect_y = [0, 0.3, 0.3]
ax.plot(rect_x, rect_y, 'k-', linewidth=2)

# 軸設定
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_aspect('equal', adjustable='box')
ax.grid(True)

# ラベルと凡例
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper left')  # 凡例を左上に配置
ax.set_title('Vector Norm and Pythagorean Theorem')

# 図を表示
plt.show()
```
