# グラム・シュミットの直交化

ここではグラム・シュミットの直交化を視覚的に表示するプログラムが書かれています。

ぜひプログラム内の値を変更してみてください。

```python
import numpy as np
import matplotlib.pyplot as plt

# Input vectors
v1 = np.array([3, 0])
v2 = np.array([2, 2])

# Calculate projection of v2 onto v1
projection = np.dot(v2, v1) / np.dot(v1, v1) * v1

# Calculate residual
residual = v2 - projection

# Plot
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v2')
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='green', label='projection')
plt.quiver(projection[0], projection[1], residual[0], residual[1], angles='xy', scale_units='xy', scale=1, color='orange', label='residual')

plt.axis('off')
plt.legend()
plt.xlim(-1, 4)
plt.ylim(-1, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
