# 固有ベクトルの図

ここでは固有ベクトルの図を視覚的に表示するプログラムが書かれています。

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義された行列
A = np.array([[2, -1],
              [-1, 2]])

# 固有値と固有ベクトルを求める
eigenvalues, eigenvectors = np.linalg.eig(A)

# グラフのサイズ設定
plt.figure(figsize=(8,8))

# 固有ベクトルを描画
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] * 3  # 固有ベクトルを3倍して描画
    plt.quiver(0, 0, vec[0].real, vec[1].real, angles='xy', scale_units='xy', scale=1, label=f'Eigenvector {i+1}', color=['green', 'purple'][i])

# グラフの範囲
plt.xlim(-2, 3)
plt.ylim(-3, 3)

# 表示
plt.grid(True)
plt.title("Grid, Column Vectors of Matrix A, and Eigenvectors")
plt.show()
```
